# **`Question 1 `: Create a minimal router with two pages: Home (/) and About (/about).Show the links to move between pages.**
---

## **`Answer -->`**


### Implementing Client-Side Routing with `react-router-dom`

> Design a basic routing architecture using a modern routing library to navigate between two functional components: `Home` and `About`. Include navigation links to allow for a seamless single-page application (SPA) experience.


### Implementation (JavaScript/JSX)

> This solution utilizes the standard industry library `react-router-dom`.

**`Code Demo`**
```javascript
import React from 'react';
import { BrowserRouter as Router, Routes, Route, Link } from 'react-router-dom';

// 1. Define Page Components
const Home = () => <h2>Home Page</h2>;
const About = () => <h2>About Page</h2>;

/**
 * App Component
 * Configures the Router and defines the Navigation structure.
 */
const App = () => {
  return (
    <Router>
      <div className="app-container">
        {/* 2. Navigation Links (Prevents Full Page Reload) */}
        <nav>
          <ul>
            <li><Link to="/">Home</Link></li>
            <li><Link to="/about">About</Link></li>
          </ul>
        </nav>

        <hr />

        {/* 3. Route Configuration */}
        <Routes>
          <Route path="/" element={<Home />} />
          <Route path="/about" element={<About />} />
        </Routes>
      </div>
    </Router>
  );
};

export default App;

```

### Technical Breakdown

#### 1. The `<Router>` Component (`BrowserRouter`)

> The `BrowserRouter` uses the HTML5 History API (pushState, replaceState, and the popstate event) to keep your UI in sync with the URL. It acts as the top-level provider for all routing information.

#### 2. `<Link>` vs. `<a>`

> In a React SPA, we use the `<Link>` component instead of standard anchor tags (`<a href="...">`).

* **Why:** Anchor tags trigger a full browser refresh, which wipes the application state. `<Link>` intercepts the click, updates the URL, and tells React to swap the components without reloading the page.

#### 3. `<Routes>` and `<Route>`

* **`<Routes>`**: Acts as a container that looks through all its child routes to find a match for the current URL.
* **`<Route>`**: Maps a specific URL `path` to a React `element`. When the URL changes to `/about`, the `About` component is rendered inside the `<Routes>` container.

#### 4. Declarative Routing

> React Router follows a **declarative** approach. Instead of manually writing logic to show or hide components based on the URL, we declare the relationship between the URL and the UI structure, allowing the library to handle the synchronization.


### Key Concepts Summary

| Component | Role |
| --- | --- |
| **`BrowserRouter`** | Provides routing context to the app. |
| **`Link`** | Navigates to a path without a page refresh. |
| **`Routes`** | A wrapper that chooses the best matching route. |
| **`Route`** | Defines the link between a path and a component. |

# **`Question 3 `: Build a DashboardLayout that shows a heading and a menu. Inside it, render DashboardHome at /dashboard and Settings at `/dashboard/settings`.**
---

## **`Answer -->`**


> To solve this, we implement a **Layout Pattern** using nested routing. This ensures that the Heading and Menu remain persistent (static) while only the inner content area (the child components) re-renders when navigating between `/dashboard` and `/dashboard/settings`.

> Here is the implementation using **React Router v6**:


### 1. The DashboardLayout Component

> The layout acts as a wrapper. The key element here is the `<Outlet />`, which serves as a placeholder for the child routes.

**`Code Demo`**
```javascript
import { Link, Outlet } from 'react-router-dom';

const DashboardLayout = () => {
  return (
    <div className="layout-wrapper">
      {/* Persistent Heading */}
      <header>
        <h1>System Dashboard</h1>
      </header>
      
      {/* Persistent Navigation Menu */}
      <nav>
        <ul>
          <li><Link to="/dashboard">Dashboard Home</Link></li>
          <li><Link to="/dashboard/settings">Settings</Link></li>
        </ul>
      </nav>

      <hr />

      {/* Dynamic Content Area */}
      <main className="content-area">
        <Outlet />
      </main>
    </div>
  );
};

```


### 2. Route Configuration

> In the main application entry point (e.g., `App.js`), we nest the routes. Notice that child paths are relative to the parent.

**`Code Demo`**
```javascript
import { BrowserRouter, Routes, Route } from 'react-router-dom';

const App = () => {
  return (
    <BrowserRouter>
      <Routes>
        {/* Parent Route providing the Layout */}
        <Route path="/dashboard" element={<DashboardLayout />}>
          
          {/* Renders at /dashboard */}
          <Route index element={<DashboardHome />} />
          
          {/* Renders at /dashboard/settings */}
          <Route path="settings" element={<Settings />} />
          
        </Route>
      </Routes>
    </BrowserRouter>
  );
};

```


### 3. Child Components (Simplified)

**`Code Demo`**

```javascript
const DashboardHome = () => <h2>Overview Metrics</h2>;
const Settings = () => <h2>Account Configurations</h2>;

```


### Technical Breakdown for Exam:

* **The `<Outlet>` Hook:** This is the most efficient way to handle nested views. It tells the router exactly where to "inject" the child component matching the current URL.
* **The `index` Route:** By using the `index` attribute, we specify that `DashboardHome` is the default component to show when the user is exactly at the parent path (`/dashboard`).
* **State Persistence:** Since the `DashboardLayout` wraps the children, the Menu and Heading do not unmount during navigation, providing a smoother User Experience (UX) and better performance.

# **`Question 4 `: Create a route /users/:userId. In the UserPage component, read the userId from the URL and display it.**

---

## **`Answer -->`**

> To solve this, we implement **Dynamic Routing**. This allows a single component to handle multiple IDs by capturing the variable part of the URL (the path parameter) and making it available to the component logic.

> Here is the implementation using **React Router v6**:


### 1. Defining the Dynamic Route

> In your `App.js` or main router configuration, use the colon (`:`) syntax to define a URL parameter. This tells the router that `userId` is a dynamic value.

**`Code Demo`**
```javascript
import { BrowserRouter, Routes, Route } from 'react-router-dom';
import UserPage from './UserPage';

const App = () => {
  return (
    <BrowserRouter>
      <Routes>
        {/* The :userId acts as a placeholder for any value */}
        <Route path="/users/:userId" element={<UserPage />} />
      </Routes>
    </BrowserRouter>
  );
};

```


### 2. The UserPage Component

> To read the parameter from the URL, we use the `useParams` hook provided by `react-router-dom`. This hook returns an object of key/value pairs of URL parameters.

```javascript
import { useParams } from 'react-router-dom';

const UserPage = () => {
  // Destructure userId from the params object
  const { userId } = useParams();

  return (
    <div className="user-profile">
      <h1>User Profile</h1>
      {/* Displaying the ID captured from the URL */}
      <p>Now viewing data for User ID: <strong>{userId}</strong></p>
    </div>
  );
};

export default UserPage;

```


### 3. Technical Breakdown for Exam:

* **Path Parameters:** By defining the path as `/users/:userId`, the router matches any URL like `/users/101` or `/users/fix-java` and assigns that specific value to the `userId` key.
* **`useParams()` Hook:** This is the standard functional approach to accessing route metadata. It automatically updates if the URL changes without unmounting the component.
* **Type Safety:** Since URL parameters are always returned as **strings**, if you need to perform calculations with the `userId`, you should use `parseInt(userId, 10)` or `Number(userId)`.


### Testing the Route:

* Navigating to `localhost:3000/users/55` will render: **Now viewing data for User ID: 55**
* Navigating to `localhost:3000/users/admin` will render: **Now viewing data for User ID: admin**

# **`Question 5 `: Only logged-in users should see /dashboard. Others should be redirected to `/login`. Write a minimal ProtectedRoute wrapper**

---

## **`Answer -->`**





> To implement a **Protected Route** (or Auth Guard), we create a Higher-Order Component (HOC) or a wrapper that evaluates the user's authentication state before rendering the requested content. If the condition is not met, we trigger a programmatic redirect to the login page.

> Here is the implementation using **React Router v6**:


### 1. The ProtectedRoute Wrapper Component

> This component checks for a boolean `isAuthenticated` (usually derived from a Global State, Context, or LocalStorage).

**`Code Demo`**

```javascript
import { Navigate } from 'react-router-dom';

const ProtectedRoute = ({ children, isAuthenticated }) => {
  // 1. Evaluate the authentication status
  if (!isAuthenticated) {
    // 2. If not logged in, redirect to /login
    // 'replace' prevents the user from navigating back to the protected route
    return <Navigate to="/login" replace />;
  }

  // 3. If logged in, render the children components
  return children;
};

export default ProtectedRoute;

```


### 2. Implementation in App.js

> We wrap the sensitive route (`/dashboard`) with the `ProtectedRoute` component.

**`Code Demo`**
```javascript
import { BrowserRouter, Routes, Route } from 'react-router-dom';

const App = () => {
  const userLoggedIn = true; // This would typically come from an Auth Provider/Context

  return (
    <BrowserRouter>
      <Routes>
        <Route path="/login" element={<LoginPage />} />
        
        {/* The Dashboard is now protected */}
        <Route
          path="/dashboard"
          element={
            <ProtectedRoute isAuthenticated={userLoggedIn}>
              <DashboardPage />
            </ProtectedRoute>
          }
        />
      </Routes>
    </BrowserRouter>
  );
};

```


### 3. Technical Breakdown for Exam:

* **The `<Navigate />` Component:** This is the declarative way to perform a redirect in React Router v6. It is preferred over `useNavigate` inside the render body to avoid side-effect errors.
* **The `replace` Prop:** This is critical for security and UX. It replaces the current entry in the history stack, so when the user is redirected to `/login`, clicking the "Back" button won't loop them back into the protected route.
* **Children Prop Pattern:** By using `{children}`, the `ProtectedRoute` becomes a reusable wrapper that can protect any component in the application without rewriting logic.
* **Conditional Rendering:** The component essentially acts as a "Gatekeeper," ensuring the `DashboardPage` component never even mounts unless the `isAuthenticated` check passes.

# **`Question 6 `: Create a counter slice with increment and decrement. Configure the store and provide it to React.**

---

## **`Answer -->`**


> To implement a state management solution using **Redux Toolkit (RTK)**, we follow a three-step process: defining the logic (Slice), setting up the container (Store), and injecting it into the UI (Provider).


### 1. Create the Slice (`counterSlice.js`)

> The slice contains the initial state and the "reducers" (functions that define how the state changes).

**`Code Demo`**
```javascript
import { createSlice } from '@reduxjs/toolkit';

const counterSlice = createSlice({
  name: 'counter',
  initialState: { value: 0 },
  reducers: {
    increment: (state) => {
      state.value += 1; // RTK uses Immer to handle immutable updates safely
    },
    decrement: (state) => {
      state.value -= 1;
    },
  },
});

export const { increment, decrement } = counterSlice.actions;
export default counterSlice.reducer;

```

### 2. Configure the Store (`store.js`)

> The store is the central repository of the application state. We gather all reducers here.

**`Code Demo`**
```javascript
import { configureStore } from '@reduxjs/toolkit';
import counterReducer from './counterSlice';

const store = configureStore({
  reducer: {
    counter: counterReducer,
  },
});

export default store;

```


### 3. Provide the Store to React (`index.js` or `main.js`)

> To make the Redux store available to the entire component tree, we wrap the App component with the `Provider`.

**`Code Demo`**
```javascript
import React from 'react';
import ReactDOM from 'react-dom/client';
import { Provider } from 'react-redux';
import store from './store';
import App from './App';

const root = ReactDOM.createRoot(document.getElementById('root'));
root.render(
  <Provider store={store}>
    <App />
  </Provider>
);

```

### Technical Breakdown for Exam:

* **`createSlice`:** This utility simplifies Redux by generating action creators and action types automatically based on the reducer names.
* **Immutability:** Unlike "vanilla" Redux, RTK allows us to write "mutating" logic (like `state.value += 1`) because it uses the **Immer** library under the hood to create a new state object.
* **The `Provider`:** This uses React's Context API to pass the store down to any nested component that needs to access state via `useSelector` or dispatch actions via `useDispatch`.
* **Store Configuration:** `configureStore` automatically sets up the Redux DevTools extension and middleware like `redux-thunk` by default.

# **`Question 7 `: Build a Counter component that shows the value and buttons to increment and decrement using Redux hooks**
---

## **`Answer -->`**

> To implement the UI for the counter, we use the **React-Redux** hooks: `useSelector` to extract data from the store and `useDispatch` to send actions to the store.


### 1. The Counter Component Implementation

> This functional component connects the Redux logic to the User Interface.

**`Code Demo`**

```javascript
import React from 'react';
// Import hooks for interacting with the Redux store
import { useSelector, useDispatch } from 'react-redux';
// Import the action creators from the slice
import { increment, decrement } from './counterSlice';

const Counter = () => {
  // 1. Access the specific state value from the store
  const count = useSelector((state) => state.counter.value);
  
  // 2. Initialize the dispatch function
  const dispatch = useDispatch();

  return (
    <div style={{ textAlign: 'center', marginTop: '20px' }}>
      <h2>Counter Application</h2>
      
      {/* 3. Display the current state */}
      <div className="display">
        <h1>{count}</h1>
      </div>

      {/* 4. Trigger actions on button clicks */}
      <div className="controls">
        <button onClick={() => dispatch(decrement())}>
          Decrement
        </button>
        
        <button onClick={() => dispatch(increment())}>
          Increment
        </button>
      </div>
    </div>
  );
};

export default Counter;

```


### 2. Technical Breakdown for Exam:

* **`useSelector` Hook:** This hook takes a selector function as an argument. The selector receives the entire Redux state and returns the specific part of the state needed (`state.counter.value`). It also subscribes the component to the store, so the component re-renders whenever that specific value changes.
* **`useDispatch` Hook:** This returns a reference to the `dispatch` function from the Redux store. It is the only way to trigger a state change.
* **Action Dispatching:** We wrap the action creators (e.g., `increment()`) inside the `dispatch()` call. This sends an action object (typically `{ type: 'counter/increment' }`) to the reducer.
* **One-Way Data Flow:** This implementation follows the strict Redux pattern:
1. **UI** triggers an event.
2. **Dispatch** sends an action.
3. **Store/Reducer** updates the state.
4. **UI** updates via the selector.

# **`Question 8 `: Create a fetchProducts thunk that loads from `/api/products`. Handle loading and error in the slice.**

---

## **`Answer -->`**




> To handle asynchronous operations in Redux Toolkit, we use the `createAsyncThunk` middleware. This allows us to manage the lifecycle of an API request (Pending, Fulfilled, and Rejected) within our slice.


### 1. Create the Asynchronous Thunk

> The thunk handles the actual side effect (the `fetch` call).



**`Code Demo`**

```javascript
import { createAsyncThunk, createSlice } from '@reduxjs/toolkit';

// Define the async thunk
export const fetchProducts = createAsyncThunk(
  'products/fetchProducts',
  async (_, { rejectWithValue }) => {
    try {
      const response = await fetch('/api/products');
      if (!response.ok) {
        throw new Error('Server Error');
      }
      const data = await response.json();
      return data;
    } catch (error) {
      // Pass the error message to the rejected action
      return rejectWithValue(error.message);
    }
  }
);

```

### 2. Handle State in the Slice (`extraReducers`)

- We use `extraReducers` to respond to the actions dispatched by the thunk.


**`Code Demo`**

```javascript
const productSlice = createSlice({
  name: 'products',
  initialState: {
    items: [],
    loading: false,
    error: null,
  },
  reducers: {}, // Standard synchronous reducers go here
  extraReducers: (builder) => {
    builder
      .addCase(fetchProducts.pending, (state) => {
        state.loading = true;
        state.error = null;
      })
      .addCase(fetchProducts.fulfilled, (state, action) => {
        state.loading = false;
        state.items = action.payload; // Data returned from the thunk
      })
      .addCase(fetchProducts.rejected, (state, action) => {
        state.loading = false;
        state.error = action.payload || 'Failed to fetch products';
      });
  },
});

export default productSlice.reducer;

```

### 3. Technical Breakdown for Exam:

* **`createAsyncThunk`:** It accepts a "type" string and a "payload creator" callback. It returns a standard Redux thunk that dispatches pending/fulfilled/rejected actions automatically.
* **`extraReducers`:** This is used to handle actions defined outside the slice (like those from an async thunk). It is the best practice for decoupling async logic from state updates.
* **The Three Promises States:**
* **Pending:** Sets `loading` to true to trigger a UI spinner.
* **Fulfilled:** Successfully populates the `items` array and disables the loader.
* **Rejected:** Captures the error message for the UI and stops the loading state.


* **`rejectWithValue`:** This utility allows us to return a custom error payload to the `rejected` case instead of a generic error object.


### Key Takeaway:

> By handling the loading and error states within the Redux store, we centralize the "Source of Truth," allowing any component in the app to know exactly when the data is fetching or if a failure occurred.

# **`Question 10 `: Create a responsive product list: on small screens one column; on medium and up, three columns. Add a hover transition. Also, show how to reuse a utility group with @apply.**

---

## **`Answer -->`**




> To implement a responsive product grid with hover effects, we leverage Tailwind CSS’s **Mobile-First breakpoint system** and the **Layer directive** for class abstraction.


### 1. Responsive Grid Implementation

> This structure uses CSS Grid utilities. By default (mobile), it occupies one column, and we use prefix modifiers (`md:`) to adjust for larger screens.


**`Code Demo`**
```html
<div class="grid grid-cols-1 md:grid-cols-3 gap-6 p-4">
  
  <div class="card-styles">
    <img src="product.jpg" alt="Product" class="w-full h-48 object-cover rounded-t-lg" />
    <div class="p-4">
      <h3 class="text-lg font-bold">Product Name</h3>
      <p class="text-gray-600">$99.00</p>
    </div>
  </div>

  </div>

```


### 2. Reusing Utilities with `@apply`

> To keep the HTML clean and ensure reusability, we can group the card's styling and hover transitions into a single CSS class inside our global stylesheet.


**`Code Demo`**
```css
/* input.css or globals.css */
@layer components {
  .card-styles {
    @apply bg-white border border-gray-200 rounded-lg shadow-sm
           transition-all duration-300 ease-in-out
           hover:shadow-xl hover:-translate-y-2 cursor-pointer;
  }
}

```


### 3. Technical Breakdown for Exam:

* **Responsive Modifiers:**
* `grid-cols-1`: The default state (0px and up).
* `md:grid-cols-3`: The breakpoint modifier (768px and up) that switches the layout to a 3-column grid.


* **Hover Transitions:**
* `transition-all`: Tells the browser to animate property changes (shadow, transform, etc.).
* `hover:-translate-y-2`: Uses a transform to lift the card slightly on the Y-axis when hovered.
* `duration-300`: Sets the animation speed to 300ms for a smooth feel.


* **The `@apply` Directive:**
* This is a Tailwind-specific CSS directive used to inline existing utility classes into custom CSS.
* **Why use it?** It prevents "class bloat" in the HTML and makes maintenance easier—if you want to change the border color for all cards, you change it in one place.
* **Placement:** It should be wrapped in `@layer components` to ensure Tailwind handles the specificity and "tree-shaking" (purging) correctly.
