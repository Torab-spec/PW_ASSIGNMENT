# **`Question 1 `: Write code that creates a new paragraph element with the text "Welcome to our website!" and adds it to a div with the id "content" on the webpage. Explain what createElement and appendChild methods do.**
---

## **`Answer -->`**


### DOM Manipulation: `createElement` and `appendChild`

- In the Document Object Model (DOM), JavaScript provides methods to dynamically modify the structure and content of a webpage. Two of the most essential methods for building content programmatically are `createElement` and `appendChild`.

#### 1. `createElement`

- The `document.createElement(tagName)` method creates a new HTML element node. It takes the tag name as a string (e.g., `'p'`, `'div'`, `'span'`) and returns a reference to that new element. **Note:** At this stage, the element exists only in memory; it is not yet visible on the webpage.

#### 2. `appendChild`

- The `node.appendChild(childNode)` method is used to add a node to the end of the list of children of a specified parent node. This method physically inserts the element into the DOM tree, causing it to render on the page.


### Implementation Example

- To create a paragraph and inject it into a `div` with the id `"content"`, follow this implementation:


<br>

**`Javascript code`**

```javascript
// 1. Create a new <p> element
const newParagraph = document.createElement("p");

// 2. Set the text content of the element
newParagraph.textContent = "Welcome to our website!";

// 3. Select the target container (div with id="content")
const contentDiv = document.getElementById("content");

// 4. Append the new paragraph to the container
contentDiv.appendChild(newParagraph);

```

#### Step-by-Step Breakdown:

* **Instantiation:** `createElement("p")` initializes an empty paragraph object in the browser's memory.
* **Content Assignment:** We set the `.textContent` property to define the visible text inside the paragraph.
* **Targeting:** `getElementById("content")` fetches the reference to the existing `div` in the DOM where we intend to place the new content.
* **Injection:** `appendChild()` links the `newParagraph` node as a child of the `contentDiv` node, finalizing the DOM update.

# **`Question 2 `: You need to change the appearance of a button when users hover over it. Write code that finds a button with id "myButton" and changes its background color to blue and text color to white using the element.style property.**

---

## **`Answer -->`**


### Modifying CSS via JavaScript: `element.style`

- In the Document Object Model (DOM), the `element.style` property provides direct access to an element's inline CSS styles. This allows you to programmatically modify the appearance of elements by setting specific CSS properties as JavaScript properties (using camelCase for hyphenated CSS names).


### Implementation Example

- To change the appearance of a button when a user hovers over it, we use the `addEventListener` method to detect the `mouseover` and `mouseout` events.

<br>

**`Javascript code`**
```javascript
// 1. Select the button element by its ID
const button = document.getElementById("myButton");

// 2. Change style on 'mouseover'
button.addEventListener("mouseover", () => {
  button.style.backgroundColor = "blue";
  button.style.color = "white";
});

// 3. Revert style on 'mouseout'
button.addEventListener("mouseout", () => {
  button.style.backgroundColor = ""; // Resets to default CSS
  button.style.color = "";           // Resets to default CSS
});

```

#### Explanation of the Code:

* **`document.getElementById`**: Retrieves a reference to the DOM element with the specific ID `"myButton"`.
* **Event Listeners**: We attach two listeners to handle the interactivity:
* `mouseover`: Triggers the style update when the pointer enters the button's area.
* `mouseout`: Reverts the styles when the pointer leaves, ensuring the button returns to its original state.


* **`element.style`**: Maps CSS properties directly to JavaScript. Note that CSS properties like `background-color` are converted to camelCase (`backgroundColor`) to conform to JavaScript variable naming conventions.

# **`Question 3 `: Create a button that shows an alert with the message "Button was clicked!" when clicked. Also, add a second event listener to the same button that changes its text to "Clicked!" after being clicked. Explain why addEventListener is better than using onclick.**

---

## **`Answer -->`**

### Event Handling: `addEventListener`

- In modern JavaScript, the `addEventListener()` method is the preferred way to attach event handlers to DOM elements. It provides a more robust and flexible system compared to the legacy `onclick` property.


### Implementation Example

- The following code demonstrates attaching multiple independent actions to a single button element.

<br>

**`Javascript code`**
```javascript
// 1. Select the button element
const btn = document.getElementById("myButton");

// 2. First event listener: Trigger an alert
btn.addEventListener("click", () => {
  alert("Button was clicked!");
});

// 3. Second event listener: Update the button text
btn.addEventListener("click", () => {
  btn.textContent = "Clicked!";
});

```

#### Why `addEventListener` is Superior to `onclick`:

1. **Multiple Listeners:** The most significant advantage is that `addEventListener` allows you to attach **multiple** event handlers for the same event type on the same element. In contrast, assigning a function to `element.onclick` will overwrite any previously defined function.
2. **Event Bubbling/Capturing Control:** `addEventListener` provides a third, optional parameter (`useCapture`) that allows you to specify whether the event should be handled during the **capturing** phase or the **bubbling** phase of the event lifecycle.
3. **Clean Separation of Concerns:** It keeps your JavaScript logic separate from your HTML. Using `onclick` often leads to "inline" event handling (e.g., `<button onclick="...">`), which violates the principle of separating structure (HTML) from behavior (JS).
4. **Removal Capabilities:** You can precisely remove a specific event listener using `removeEventListener()`. With `onclick`, you would have to manually reset the property to `null` or an empty function, which lacks granular control.

# **`Question 4 `: Write a function named extractDomain that takes an email string like `"john@example.com"` and returns the domain part `("example.com")` using only string methods. Do not use regular expressions.**
---

## **`Answer -->`**


### String Manipulation: Domain Extraction

- To extract a substring from an email address without using regular expressions, we leverage the `indexOf()` method to locate the separator character (`@`) and the `slice()` or `substring()` method to extract the portion of the string that follows it.


### Implementation Example

- The following code calculates the position of the `@` symbol and extracts the remaining characters to isolate the domain.

<br>

**`Javascript code`**
```javascript
/**
 * Extracts the domain from an email address string.
 * @param {string} email - The full email address.
 * @returns {string} The domain portion.
 */
function extractDomain(email) {
  // 1. Find the index of the '@' character
  const atIndex = email.indexOf('@');
  
  // 2. Return the substring starting one character after the '@'
  // Adding 1 skips the '@' symbol itself
  return email.slice(atIndex + 1);
}

// Usage:
console.log(extractDomain("john@example.com")); // Output: "example.com"
console.log(extractDomain("dev.team@company.org")); // Output: "company.org"

```

#### Explanation of the Logic:

1. **`indexOf('@')`**: This method searches the string from left to right and returns the integer index of the first occurrence of the `@` character.
2. **`slice(atIndex + 1)`**: The `slice` method extracts a section of a string and returns it as a new string. By providing `atIndex + 1` as the argument, we instruct the engine to start copying from the character immediately following the `@` symbol until the end of the string.
3. **Efficiency**: This approach is highly performant as it performs a linear scan to locate the index and a single memory copy operation for the slice.

# **`Question 5 `: Write a function called repeatMessage that takes a message string and a number, and returns the message repeated that many times, separated by a space. Do not use loops.**



<h2>
<b>
<code>
Example:
repeatMessage("hello", 3) should return "hello hello hello"
</code>

---

## **`Answer -->`**

### String Manipulation: Array Methods for Repetition

- To achieve string repetition without using loops, we can leverage built-in array prototype methods. This approach is highly efficient and declarative, following functional programming patterns.


### Implementation Example

- The optimal way to solve this in modern JavaScript is to create an empty array of a specific length, fill it with the desired message, and then join the elements together.

<br>

**`Javascript code`**
```javascript
/**
 * Repeats a message string a specific number of times.
 * @param {string} message - The string to repeat.
 * @param {number} count - The number of times to repeat the message.
 * @returns {string} The repeated string separated by spaces.
 */
function repeatMessage(message, count) {
  // 1. Create an array of 'count' length
  // 2. Fill the array with the 'message' string
  // 3. Join the elements with a space
  return Array(count).fill(message).join(' ');
}

// Usage:
console.log(repeatMessage("hello", 3)); // Output: "hello hello hello"

```

#### Explanation of the Code:

1. **`Array(count)`**: This initializes a new `Array` object with a length equal to `count`. At this stage, the array contains "empty slots" (holes).
2. **`.fill(message)`**: The `fill()` method populates every index of the array with the provided `message` string. This effectively creates an array like `["hello", "hello", "hello"]`.
3. **`.join(' ')`**: The `join()` method concatenates all elements of the array into a single string, inserting the specified separator (a space) between each element.

# **`Question 6 `: Explain what the Event Loop is in JavaScript and predict the output of the following code. Also explain why the output appears in that specific order.**

<h2>
<b>
<code>
console.log('Start');<br>
setTimeout(function() {<br>
console.log('Timeout 1');<br>
}, 0);<br>
console.log('Middle');<br>
setTimeout(function() {<br>
console.log('Timeout 2');<br>
}, 0);<br>
console.log('End');<br>
</code>
</b>
</h2>

---

## **`Answer -->`**

### The JavaScript Event Loop

- The **Event Loop** is a foundational mechanism that enables JavaScript's non-blocking, asynchronous behavior, despite being single-threaded. Because the JavaScript engine can only execute one task at a time, it uses the Event Loop to coordinate the execution of code, process events, and manage the execution of queued sub-tasks.

#### Core Components:

1. **Call Stack:** Where synchronous code is executed. Functions are pushed onto the stack and popped off once complete.
2. **Web APIs:** Browser-provided environments (like `setTimeout` or DOM events) where asynchronous tasks are offloaded.
3. **Callback Queue (Task Queue):** Where completed asynchronous tasks (callbacks) wait to be pushed onto the Call Stack.

The Event Loop constantly monitors the Call Stack. When the stack is empty, it checks the Callback Queue. If tasks are waiting, it pushes the first task from the queue into the Call Stack for execution.


### Code Execution Analysis

<br>

**`Javascript code`**
```javascript
console.log('Start');

setTimeout(function() {
  console.log('Timeout 1');
}, 0);

console.log('Middle');

setTimeout(function() {
  console.log('Timeout 2');
}, 0);

console.log('End');

```

#### `Predicted Output:`

```text
Start
Middle
End
Timeout 1
Timeout 2

```


### Why this order occurs:

1. **Synchronous Execution:** The engine executes code line-by-line from top to bottom. `console.log('Start')`, `console.log('Middle')`, and `console.log('End')` are pushed onto the **Call Stack** and executed immediately.
2. **Asynchronous Offloading:** When the engine encounters `setTimeout`, it hands the timer off to the **Web API**. Even with a delay of `0ms`, the Web API schedules these functions to be pushed into the **Callback Queue** only after the timer expires.
3. **Event Loop Priority:** The Event Loop does not move anything from the Callback Queue to the Call Stack until the **Call Stack is entirely empty**.
4. **Queue Processing:** Since all synchronous code (`Start`, `Middle`, `End`) finishes first, the Call Stack clears. The Event Loop then sees `Timeout 1` and `Timeout 2` in the Callback Queue. It processes them in the order they were added (First-In, First-Out).

# **`Question 7 `: Write a JavaScript function named startCountdown that logs “Time left:X” every second, starting from 5 down to 1, and finally logs “Time’s up!”. UsesetInterval and clearInterval only.**

---

## **`Answer -->`**


### JavaScript Timer Implementation: `setInterval` and `clearInterval`

- To create a countdown timer, we use `setInterval`, which repeatedly executes a function at a fixed time interval. We pair this with a conditional check to clear the interval using `clearInterval` once the target condition is met, preventing the timer from running indefinitely.


### Implementation Example

<br>

**`Javascript code`**
```javascript
/**
 * Initiates a countdown timer from 5 to 1.
 */
function startCountdown() {
  let counter = 5;

  // 1. Set the interval to run every 1000 milliseconds (1 second)
  const timerId = setInterval(() => {
    if (counter > 0) {
      console.log(`Time left: ${counter}`);
      counter--;
    } else {
      // 2. Clear the interval when counter hits 0
      clearInterval(timerId);
      console.log("Time’s up!");
    }
  }, 1000);
}

// Invoke the function
startCountdown();

```

#### Explanation of the Code:

1. **State Initialization:** We define a `counter` variable to track the remaining time.
2. **`setInterval(callback, 1000)`**: This registers a periodic task that executes the provided arrow function every $1,000$ milliseconds. It returns a unique numeric ID (`timerId`), which serves as a reference to that specific interval.
3. **Conditional Logic:** Inside the callback, we check the `counter` value.
* If it is greater than $0$, we log the current value and decrement.
* If the counter reaches $0$, we execute `clearInterval(timerId)`. This stops the interval from triggering again, effectively halting the execution cycle.


4. **Termination:** Once cleared, the function logs the final message, "Time’s up!", signaling the end of the process.

# **`Question 8 `: Explain what Promises are in JavaScript and write code that creates a Promise to simulate checking if a user is logged in. The Promise should resolve with user data after 2 seconds if successful, or reject with an error message if the user is not found.**


---

## **`Answer -->`**


### JavaScript Promises

- A **Promise** in JavaScript is an object representing the eventual completion (or failure) of an asynchronous operation and its resulting value. It acts as a placeholder for a value that is not yet known but will be available in the future.

#### The Three States of a Promise:

* **Pending:** The initial state; the operation has not finished yet.
* **Fulfilled (Resolved):** The operation completed successfully, and the Promise has a resulting value.
* **Rejected:** The operation failed, and the Promise has a reason for the failure (usually an error object).


### Implementation Example

- The following code simulates an asynchronous check (e.g., a network request to an authentication server) using the `Promise` constructor.

<br>

**`Javascript code`**
```javascript
/**
 * Simulates checking user authentication status.
 * @param {boolean} isLoggedIn - Flag to simulate success or failure.
 * @returns {Promise<string>}
 */
function checkUserLogin(isLoggedIn) {
  return new Promise((resolve, reject) => {
    console.log("Checking user status...");

    // Simulate an asynchronous operation (e.g., API call) with a 2-second delay
    setTimeout(() => {
      if (isLoggedIn) {
        resolve({ username: "Fix java", status: "Active" });
      } else {
        reject("Error: User not found or not logged in.");
      }
    }, 2000);
  });
}

// Consuming the Promise
checkUserLogin(true)
  .then((data) => {
    console.log("Success:", data);
  })
  .catch((error) => {
    console.error(error);
  });

```

#### Explanation of the Code:

1. **The Executor Function:** The `Promise` constructor takes a callback function called an *executor*, which receives two arguments: `resolve` (to mark success) and `reject` (to mark failure).
2. **`setTimeout`:** This mimics a real-world delay (like waiting for a database query or network response).
3. **`resolve` vs `reject`:** If `isLoggedIn` is true, we call `resolve()` and pass the user data. If false, we call `reject()` with an error message.
4. **Chaining:** We use `.then()` to handle the successfully resolved value and `.catch()` to handle any rejections or errors that occur during the process.

# **`Question 9 `: Rewrite the following Promise-based code to use async/await syntax. The code fetches user data and then fetches that user's posts. Include proper error handling.**

<h2>
<b>
<code>
function getUserDataAndPosts(userId) {<br>
return fetchUserData(userId)<br>
.then(userData => {<br>
return fetchUserPosts(userData.id)<br>
.then(posts => {<br>
return { user: userData, posts: posts };<br>
});<br>
})<br>
.catch(error => {<br>
console.error('Error:', error);<br>
throw error;<br>
});<br>
}
</code>
</b>
</h2>

---

## **`Answer -->`**

### Refactoring Promises to `async/await`

- The `async/await` syntax, introduced in ES2017, provides a cleaner, more readable way to work with asynchronous code by allowing it to be written in a style that resembles synchronous execution.

* **`async`**: Declares that a function will handle asynchronous operations and automatically returns a Promise.
* **`await`**: Pauses the execution of the function until the Promise settles (resolves or rejects), making the flow of data explicit.


### Implementation Example

- By using `async/await`, we eliminate the need for deeply nested `.then()` callbacks, often referred to as "callback hell." Proper error handling is implemented using a standard `try...catch` block.

<br>

**`Javascript code`**

```javascript
/**
 * Fetches user data and their posts using async/await.
 * @param {number|string} userId - The unique identifier for the user.
 * @returns {Promise<Object>} An object containing user data and their posts.
 */
async function getUserDataAndPosts(userId) {
  try {
    // 1. Fetch user data
    const userData = await fetchUserData(userId);
    
    // 2. Fetch user's posts using the ID from the previous request
    const posts = await fetchUserPosts(userData.id);
    
    // 3. Return the combined data
    return { user: userData, posts: posts };
    
  } catch (error) {
    // 4. Handle errors from either fetch call
    console.error('Error fetching data:', error);
    throw error; // Rethrow to allow the caller to handle the failure
  }
}

```

#### Key Improvements:

1. **Readability:** The code executes line-by-line, which is easier to follow than chained `.then()` blocks.
2. **Scope management:** Variables like `userData` remain accessible in the subsequent lines of the function scope, removing the need to pass them down into inner callbacks.
3. **Unified Error Handling:** The `try...catch` block centralizes error handling, catching exceptions from both `fetchUserData` and `fetchUserPosts` in a single location rather than needing separate `.catch()` handlers for each step.

### Fetching Data with the Fetch API

- The `fetch()` API provides a modern, promise-based interface for making network requests. When interacting with public APIs, it is standard practice to handle the response using `async/await` syntax for better readability and to ensure that potential errors (such as network failures or 404 responses) are caught and handled gracefully.


### Implementation Example

- The following code fetches data from the OpenWeatherMap API and dynamically updates an HTML element to display the weather.

<br>

**`Javascript code`**

```javascript
const apiKey = 'abcd1234';
const apiUrl = `https://api.openweathermap.org/data/2.5/weather?q=Mumbai&appid=${apiKey}&units=metric`;

async function getWeatherData() {
  try {
    // 1. Initiate the fetch request
    const response = await fetch(apiUrl);

    // 2. Check if the response status is successful
    if (!response.ok) {
      throw new Error(`HTTP error! status: ${response.status}`);
    }

    // 3. Parse the JSON response
    const data = await response.json();

    // 4. Extract and display data
    displayWeather(data);
    
  } catch (error) {
    console.error("Could not fetch weather data:", error);
    document.getElementById("weather-display").innerHTML = "Failed to load weather.";
  }
}

function displayWeather(data) {
  const displayElement = document.getElementById("weather-display");
  
  // Format the output
  displayElement.innerHTML = `
    <h3>Weather in ${data.name}</h3>
    <p>Temperature: ${data.main.temp}°C</p>
    <p>Condition: ${data.weather[0].description}</p>
    <p>Humidity: ${data.main.humidity}%</p>
  `;
}

// Call the function
getWeatherData();

```

#### Key Considerations:

* **Response Validation:** Always check `response.ok`. A `fetch()` request only rejects a promise if there is a network failure; it will *not* reject on HTTP error statuses (like 404 or 500) by default.
* **JSON Parsing:** The `response.json()` method is asynchronous and must be awaited to retrieve the actual data object.
* **Error Handling:** The `try...catch` block ensures that any issues during the fetch, parsing, or DOM manipulation are caught and logged, preventing the application from crashing.
