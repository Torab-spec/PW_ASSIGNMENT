# **`Question 2` : Write a function checkEligibility that takes an age as a parameter and returns:**

- ## **"Eligible" if age is 18 or above**
- ## **"Not Eligible" if age is below 18**

# **Use an if-else statement.**

## **`ANSWER -->`**


- Writing a JavaScript function named `checkEligibility` that accepts `age` as a parameter. The function should return "Eligible" if the age is 18 or older, and "Not Eligible" if the age is under 18, using an `if-else` statement.


### JavaScript Code Solution:

<p align="right"><code>JavaScript</code></p>

```javascript
function checkEligibility(age) {
    if (age >= 18) {
        return "Eligible";
    } else {
        return "Not Eligible";
    }
}

```


### Explanation:

* **Function Declaration:** We define a function named `checkEligibility` that takes one argument, `age`.
* **Condition:** The `if` statement checks if the value of `age` is greater than or equal to `18` ($age \ge 18$).
* **Return Values:** * If the condition is **true**, the function executes the first block and returns the string `"Eligible"`.
* If the condition is **false** (meaning the age is 17 or lower), the `else` block executes and returns `"Not Eligible"`.




### Example Usage:

<p align="right"><code>JavaScript</code></p>

```javascript
console.log(checkEligibility(20)); // Output: "Eligible"
console.log(checkEligibility(15)); // Output: "Not Eligible"

```

# **`Question 3 `: Write a function getDayName that takes a number (1-7) and returns the corresponding day of the week. Use a switch statement.**

## **`ANSWER -->`**


- Writing a JavaScript function named `getDayName` that takes a number (1 to 7) as a parameter and returns the corresponding day of the week (e.g., 1 for Monday, 2 for Tuesday, etc.) using a `switch` statement.


### JavaScript Code Solution:

<p align="right"><code>JavaScript</code></p>

```
function getDayName(dayNumber) {
    let day;

    switch (dayNumber) {
        case 1:
            day = "Monday";
            break;
        case 2:
            day = "Tuesday";
            break;
        case 3:
            day = "Wednesday";
            break;
        case 4:
            day = "Thursday";
            break;
        case 5:
            day = "Friday";
            break;
        case 6:
            day = "Saturday";
            break;
        case 7:
            day = "Sunday";
            break;
        default:
            day = "Invalid day number. Please enter a number between 1 and 7.";
    }

    return day;
}

```


### Key Components of the Switch Statement:

* **`switch (dayNumber)`**: This evaluates the input variable `dayNumber`.
* **`case`**: Each case compares the input to a specific value (1 through 7).
* **`break`**: This keyword is essential; it stops the execution from "falling through" to the next case once a match is found.
* **`default`**: This block runs if the input does not match any of the defined cases (e.g., if the user enters 10).
* **`return day`**: After the switch block finishes, the function returns the assigned string.


### Example Usage:


<p align="right"><code>JavaScript</code></p>

```
console.log(getDayName(1)); // Output: "Monday"
console.log(getDayName(5)); // Output: "Friday"
console.log(getDayName(9)); // Output: "Invalid day number..."

```

# **`Question 5 `: Given an array numbers = [1, 2, 3, 4, 5], perform the following:**
- ## **Use a for loop to print each number.**
- ## **Use a while loop to print each number.**
- ## **Use a for-in loop to print each index.**
- ## **Use a for-of loop to print each number.**

## **`ANSWER -->`**

- Given an array `numbers = [1, 2, 3, 4, 5]`, demonstrate four different types of loops to iterate through the array: a standard `for` loop, a `while` loop, a `for-in` loop, and a `for-of` loop.

---

### Step 1: Standard `for` loop

The standard `for` loop uses an index variable to access each element. It consists of initialization, a condition, and an increment.

<p align="right"><code>JavaScript</code></p>

```javascript
const numbers = [1, 2, 3, 4, 5];

for (let i = 0; i < numbers.length; i++) {
    console.log(numbers[i]);
}
// Output: 1, 2, 3, 4, 5

```

### Step 2: `while` loop

The `while` loop continues to execute as long as the specified condition evaluates to `true`. You must manually handle the counter/index.

<p align="right"><code>JavaScript</code></p>

```javascript
let i = 0;
while (i < numbers.length) {
    console.log(numbers[i]);
    i++;
}
// Output: 1, 2, 3, 4, 5

```

### Step 3: `for-in` loop (Printing Indexes)

The `for-in` loop is designed to iterate over the **enumerable properties** (keys/indexes) of an object or array.

<p align="right"><code>JavaScript</code></p>

```javascript
for (let index in numbers) {
    console.log(index);
}
// Output: "0", "1", "2", "3", "4"

```

*Note: In JavaScript, array indexes are returned as strings during `for-in` iteration.*

### Step 4: `for-of` loop (Printing Values)

The `for-of` loop is the modern way to iterate directly over the **values** of an iterable object (like an array).

<p align="right"><code>JavaScript</code></p>

```javascript
for (let num of numbers) {
    console.log(num);
}
// Output: 1, 2, 3, 4, 5

```


### Summary Table: Loop Comparison

| Loop Type | Primary Use Case | Accesses |
| --- | --- | --- |
| **`for`** | General purpose / precise control | Index ($i$) |
| **`while`** | When number of iterations is unknown | Index ($i$) |
| **`for-in`** | Iterating over keys/properties | Index (Key) |
| **`for-of`** | Iterating over values directly | Element (Value) |

## **`Question 6` : Explain what functions are in JavaScript and provide an example of a function that adds two numbers.**


 ## **`ANSWER -->`**

 ### Functions in JavaScript

A **function** in JavaScript is a reusable block of code designed to perform a specific task. Instead of writing the same logic multiple times, you define it once in a function and "call" or "invoke" it whenever needed.

Think of a function like a **recipe**:

1. **Input (Arguments):** The ingredients you put in.
2. **Process (Body):** The steps taken to cook the meal.
3. **Output (Return Value):** The finished dish.


### Key Components of a Function:

* **Function Keyword:** Used to declare a function.
* **Name:** A unique identifier (e.g., `addNumbers`).
* **Parameters:** Variables listed inside parentheses `()` that act as placeholders for input values.
* **Function Body:** The code enclosed in curly braces `{}` that executes the logic.
* **Return Statement:** Used to send a value back to where the function was called.


### Example: A Function to Add Two Numbers

Here is a standard example of a function that takes two numbers as inputs and returns their sum.

// 1. Function Declaration
function addTwoNumbers(num1, num2) {
    // 2. Logic: Adding the parameters
    let sum = num1 + num2;

    // 3. Returning the result
    return sum;
}

// 4. Function Call (Invoking the function)
let result = addTwoNumbers(10, 5);

console.log(result); // Output: 15

```

### Explanation of the Example:

* **`num1` and `num2**` are parameters. When we call the function with `(10, 5)`, `num1` becomes 10 and `num2` becomes 5.
* The **`return`** keyword is crucial; without it, the function would perform the math but "keep" the answer to itself. Returning allows us to store the result in the `result` variable.


### Why Use Functions?

* **DRY (Don't Repeat Yourself):** Reduces code duplication.
* **Organization:** Breaks complex programs into smaller, manageable pieces.
* **Maintainability:** If you need to change the logic, you only have to change it in one place.

---
---

---
---
# **`Question 7 `: Convert the add function from Question 6 into an arrow function**

## **`ANSWER -->`**


- Rewrite the `addTwoNumbers` function (from Question 6) using the **Arrow Function** syntax introduced in ES6.


### JavaScript Code Solution:
<p align="right"><code>JavaScript</code></p>


```javascript
const addTwoNumbers = (num1, num2) => {
    return num1 + num2;
};

```

**Shorthand Version (Implicit Return):**
If a function body contains only a single expression, you can omit the curly braces `{}` and the `return` keyword for a more concise syntax:

<p align="right"><code>JavaScript</code></p>

```javascript
const addTwoNumbers = (num1, num2) => num1 + num2;

```


### Key Differences from Regular Functions:

* **Syntax:** Arrow functions use the `=>` (fat arrow) operator instead of the `function` keyword.
* **Assignment:** They are usually assigned to a variable (using `const` or `let`) because they are anonymous by nature.
* **Conciseness:** As shown in the shorthand version, they allow for very "clean" code when performing simple operations.
* **`this` Binding:** Unlike regular functions, arrow functions do not have their own `this` context (they inherit it from the surrounding code), which is a major advantage in certain programming scenarios.


### Example Usage:

<p align="right"><code>JavaScript</code></p>


```javascript
console.log(addTwoNumbers(12, 8)); // Output: 20

```

# **`Question 8` : Write a function greet that takes a name and a callback function. The greet function should call the callback with the name.**

## **`ANSWER -->`**


- Writing a JavaScript function named `greet` that takes two parameters: a `name` (string) and a `callback` (a function). The `greet` function should then invoke the provided callback function, passing the `name` as an argument.


### JavaScript Code Solution:
<p align="right"><code>JavaScript</code></p>

```javascript
/**
 * @param {string} name - The name to be greeted.
 * @param {function} callback - The function to be executed.
 */
function greet(name, callback) {
    // Invoking the callback function and passing the name parameter to it
    callback(name);
}

// Example of a callback function to use with greet
function logGreeting(userName) {
    console.log("Hello, " + userName + "!");
}

// Calling the greet function
greet("Rahul", logGreeting);
// Output: Hello, Rahul!

```

### Step-by-Step Explanation:

1. **Function Definition**: The `greet` function is defined with two parameters. The second parameter, `callback`, is expected to be a function.
2. **Higher-Order Function**: In JavaScript, functions that accept other functions as arguments are called **Higher-Order Functions**.
3. **Function Execution**: Inside the body of `greet`, we use the syntax `callback(name)`. This tells JavaScript to execute whatever function was passed in and provide the `name` variable to it.
4. **Flexibility**: This pattern allows the `greet` function to remain flexible. Instead of just logging to the console, the callback could save the name to a database, alert it in the browser, or transform the text.


### Alternative: Using an Anonymous Arrow Function

In exams, you may also show how to call the function using an anonymous arrow function for a more modern approach:
<p align="right"><code>JavaScript</code></p>

```javascript
greet("Amit", (n) => {
    console.log("Welcome, " + n);
});

```

# **Question 9: Given the array:**
# **const fruits = `["apple", "banana", "cherry"]`;**
- ## **What is the length of the array?**
- ## **Add "date" to the end of the array using the length property**


## **`ANSWER -->`**

- Given the array `const fruits = ["apple", "banana", "cherry"];`, determine the length of the array and add the string `"date"` to the end of the array using the `length` property (instead of the `push()` method).


### JavaScript Code Solution:
<p align="right"><code>JavaScript</code></p>

```
const fruits = ["apple", "banana", "cherry"];

// 1. Finding the length of the array
const arrayLength = fruits.length;
console.log(arrayLength); // Output: 3

// 2. Adding "date" to the end using the length property
fruits[fruits.length] = "date";

console.log(fruits); // Output: ["apple", "banana", "cherry", "date"]

```


### Step-by-Step Explanation:

1. **Determining the Length**: The `.length` property returns the total number of elements currently in the array. Since there are three elements ("apple", "banana", "cherry"), `fruits.length` is **3**.
2. **Array Indexing**: In JavaScript, arrays are **zero-indexed**, meaning the first element is at index 0, the second at index 1, and the third at index 2.
3. **The Logic of Using `.length**`: Because the index of the last element is always `length - 1`, the value of the `length` property itself points to the **next available slot** at the end of the array.
* In this case, `fruits.length` is 3.
* By assigning `fruits[3] = "date"`, we occupy that new slot.


4. **Result**: The array is updated dynamically, and the new length automatically becomes **4**.


### Final Exam Summary:

* **Initial Length:** 3
* **Method Used:** Array index assignment via `array[array.length]`
* **Updated Array:** `["apple", "banana", "cherry", "date"]`

# **`Question 10` : Given the string text = " Hello World! ", perform the following:**
- ## **Trim the whitespace.**
- ## **Convert to uppercase.**
- ## **Check if it includes "World"**

## **`ANSWER-->`**


- Given the string `let text = " Hello World! ";`, perform three operations: remove the leading and trailing whitespace, convert the entire string to uppercase, and check if the string contains the word "World".


### JavaScript Code Solution:

<p align="right"><code>JavaScript</code></p>

```
let text = " Hello World! ";

// Step 1: Trim the whitespace
let trimmedText = text.trim();
console.log(trimmedText); // Output: "Hello World!"

// Step 2: Convert to uppercase
let upperText = trimmedText.toUpperCase();
console.log(upperText); // Output: "HELLO WORLD!"

// Step 3: Check if it includes "World"
let hasWorld = trimmedText.includes("World");
console.log(hasWorld); // Output: true

```


### Step-by-Step Explanation:

**1. Trimming Whitespace (`.trim()`)**

* **Task:** Remove the empty spaces at the beginning and the end of the string.
* **Explanation:** The `trim()` method creates a new string without modifying the original. It removes spaces, tabs, and newline characters from both ends.
* **Result:** `" Hello World! "` becomes `"Hello World!"`.

**2. Converting to Uppercase (`.toUpperCase()`)**

* **Task:** Change all alphabetical characters in the string to their capital forms.
* **Explanation:** The `toUpperCase()` method is used. Note that non-alphabetical characters like the exclamation mark (`!`) remain unchanged.
* **Result:** `"Hello World!"` becomes `"HELLO WORLD!"`.

**3. Checking for a Substring (`.includes()`)**

* **Task:** Verify if a specific sequence of characters exists within the string.
* **Explanation:** The `includes()` method returns a **Boolean** value (`true` or `false`). It is case-sensitive, meaning it would return `false` if you searched for `"world"` instead of `"World"`.
* **Result:** Since `"World"` is present in the trimmed string, the result is `true`.


### Final Summary Table:

| Operation | Method | Resulting Value |
| --- | --- | --- |
| **Trim** | `.trim()` | `"Hello World!"` |
| **Uppercase** | `.toUpperCase()` | `"HELLO WORLD!"` |
| **Search** | `.includes("World")` | `true` |

# **`Question 11 `: Given the array `numbers = [1, 2, 3, 4, 5],` perform the following:**
- ## **Add 6 to the end.**
- ## **Remove the first element.**
- ## **Find the index of 4.**
- ## **Check if all numbers are greater than 0**

## **`ANSWER -->`**


- Given the array `let numbers = [1, 2, 3, 4, 5];`, perform the following operations: add 6 to the end, remove the first element, find the index of 4, and check if all elements are greater than 0.


### Step-by-Step Solutions:

#### 1. Add 6 to the end

We use the `.push()` method, which adds one or more elements to the end of an array and returns the new length of the array.
<p align="right"><code>JavaScript</code></p>

```javascript
numbers.push(6);
console.log(numbers); // Output: [1, 2, 3, 4, 5, 6]

```

#### 2. Remove the first element

We use the `.shift()` method, which removes the first element from an array and returns that removed element.
<p align="right"><code>JavaScript</code></p>

```javascript
numbers.shift();
console.log(numbers); // Output: [2, 3, 4, 5, 6]

```

#### 3. Find the index of 4

We use the `.indexOf()` method, which returns the first index at which a given element can be found in the array.
<p align="right"><code>JavaScript</code></p>

```javascript
let index = numbers.indexOf(4);
console.log(index); // Output: 2

```

*Note: Because we removed the 1, the new array is `[2, 3, 4, 5, 6]`. The value `4` is at index 2.*

#### 4. Check if all numbers are greater than 0

We use the `.every()` method. This tests whether all elements in the array pass the test implemented by the provided function.
<p align="right"><code>JavaScript</code></p>

```javascript
let allPositive = numbers.every(num => num > 0);
console.log(allPositive); // Output: true

```


### Consolidated Code Demonstration:
<p align="right"><code>JavaScript</code></p>

```javascript
let numbers = [1, 2, 3, 4, 5];

// 1. Add 6 to the end
numbers.push(6);

// 2. Remove the first element
numbers.shift();

// 3. Find index of 4
let index = numbers.indexOf(4);

// 4. Check if all are > 0
let allPositive = numbers.every(num => num > 0);

console.log("Final Array:", numbers);     // [2, 3, 4, 5, 6]
console.log("Index of 4:", index);       // 2
console.log("All > 0?:", allPositive);   // true

```


### Summary Table for Exam Review:

| Operation | Method | Description |
| --- | --- | --- |
| **Add to end** | `.push()` | Appends element to the end. |
| **Remove from start** | `.shift()` | Removes the first element. |
| **Find index** | `.indexOf()` | Returns index of the first occurrence. |
| **Validation** | `.every()` | Returns `true` if all elements pass the test. |
