# **Question 1 : Create a layout with three boxes side by side using Flexbox. The first box should have twice the width of the other two. Provide the HTML and CSS code snippets.**

## **ANSWER :**

- Here is the HTML and CSS code to create three side-by-side boxes where the first box is twice the width of the other two.

**`HTML(index.html) `**
```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <title>Q1 | Flexbox Layout</title>
    <link rel="stylesheet" href="style.css" />
  </head>
  <body>
    <div class="container">
      <div class="box box-1">Box 1 (Twice the width)</div>

      <div class="box box-2">Box 2</div>

      <div class="box box-3">Box 3</div>
    </div>
  </body>
</html>
```
**`CSS(style.css)`**
```
/* 1. Set up the container as a Flexbox.*/
* {
  padding: 0;
  margin: 0;
  box-sizing: border-box;
}

.container {
  /* Enables flex context for all direct children */
  display: flex;
  /* Just to center the layout on the page */
  width: 90%;
  max-width: 800px;
  margin: 40px auto;
  border: 2px solid #333; /* For visualization */
  min-height: 150px;
}

/* 2. Style the individual boxes (flex-items).*/

.box {
  /* flex-grow, flex-shrink, flex-basis:
    flex: 1 1 0; is the shorthand for "flex: 1" */
  flex: 1;
  padding: 20px;
  text-align: center;
  color: white;
  font-size: 1.2em;
  font-family: sans-serif;
}

/* 3. The Key Step: Set the flex-grow value for the first box to 2.
       This tells the browser to allocate twice as much *remaining space* to this box compared to the others (which default to 1).
*/
.box-1 {
  flex: 2; /* Allocates 2 parts of the available space */
  background-color: #e74c3c; /* Red */
}

/* Style for the normal-width boxes (flex: 1) */
.box-2 {
  flex: 1; /* Allocates 1 part of the available space */
  background-color: #2ecc71; /* Green */
}

.box-3 {
  flex: 1; /* Allocates 1 part of the available space */
  background-color: #3498db; /* Blue */
}

```

# **Question 2: Using CSS Grid, create a 2x2 grid layout where the first row spans two columns, and the second row has two equal columns. Provide the HTML and CSS code snippets**

## **ANSWER :**

- We can create the desired 2x2 grid layout using CSS Grid with the following HTML and CSS. The code snippets is below ðŸ‡°

`HTML(index.html)`

```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Q2 |</title>
    <link rel="stylesheet" href="style.css" />
  </head>
  <body>
    <div class="grid-container">
      <div class="grid-item item-1">Item 1 (Spans 2 columns)</div>
      <div class="grid-item item-2">Item 2</div>
      <div class="grid-item item-3">Item 3</div>
    </div>
  </body>
</html>
```

`CSS(style.css)`

```
* {
  padding: 0;
  margin: 0;
  box-sizing: border-box;
}

.grid-container {
  display: grid;
  /* Define 2 equal-width columns */
  grid-template-columns: 1fr 1fr;
  /* Define 2 rows. Auto is used here, but specific heights could be used (e.g., 100px 100px) */
  grid-template-rows: auto auto;
  gap: 10px; /* Space between grid items */
  width: 400px; /* Example width for the container */
  border: 1px solid #ccc;
  padding: 10px;
}

.grid-item {
  background-color: #f0f0f0;
  border: 1px solid #333;
  padding: 15px;
  text-align: center;
}

/* Make the first item span two columns for the first row */
.item-1 {
  /* Starts at column line 1 and ends before column line 3, spanning 2 columns */
  grid-column: 1 / 3;
  /* Alternatively: grid-column: span 2; */
  /* Ensure it's in the first row */
  grid-row: 1;
  background-color: lightblue;
}

/* Items 2 and 3 naturally fall into the second row and occupy one column each,
   as the grid is defined with 2 columns and item 1 took up all the space in row 1. */
.item-2 {
  grid-row: 2;
  grid-column: 1;
}

.item-3 {
  grid-row: 2;
  grid-column: 2;
}
```

# **Question 4: Create a button that changes its background color to blue and scales up slightly when hovered over, using CSS transitions. Provide the HTML and CSS code snippets.**

## **ANSWER :**

- Here are the HTML and CSS code snippets to create a button that changes its background color and scales up on hover using CSS transitions.

`HTML(index.html)`

```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Q-4</title>
    <link rel="stylesheet" href="style.css" />
  </head>
  <body>
    <button class="transition-button">Hover Me</button>
  </body>
</html>
```

`CSS(style.css)`

```
* {
  padding: 0;
  margin: 0;
  box-sizing: border-box;
}
body {
  height: 100vh;
  width: 100vw;
  display: flex;
  justify-content: center;
  align-items: center;
}

.transition-button {
  display: inline-block;
  /* Initial Styles */
  padding: 15px 25px;
  font-size: 16px;
  color: white;
  background-color: #72bff3; /* Default color */
  border: none;
  border-radius: 5px;
  outline: none;

  /* Set up the transition */

  transition: background-color 0.3s ease-in-out, transform 0.2s ease-in-out;
}

.transition-button:hover {
  /* Hover Styles */
  background-color: #2980b9; /* Darker blue on hover */
  transform: scale(1.1); /*Scaling on hover*/
}
```

# **Question 6: Create a simple animation that moves a square box from left to right infinitely. Provide the HTML and CSS code snippets.**

## **ANSWER :**
- We can create a simple, infinite left-to-right animation of a square box using HTML and CSS with the @keyframes rule.

- Code snippets are below :

`HTML(index.html)`

```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Q-6</title>
    <link rel="stylesheet" href="style.css" />
  </head>
  <body>
    <div class="box-container">
      <div class="moving-box"></div>
    </div>
  </body>
</html>
```

`CSS(style.css)`
```
* {
  padding: 0;
  margin: 0;
  box-sizing: border-box;
}
/* Basic reset and container styling for context */
body {
  margin: 0;
  padding: 0;
  height: 100vh;
  display: flex;
  align-items: center;
  justify-content: center;
  background-color: #f0f0f0;
}

.box-container {
  width: 80%;
  height: 100px;
  border: 2px solid #ccc;
  position: relative;
  overflow: hidden;
}

/* The moving element */
.moving-box {
  width: 50px;
  height: 50px;
  background-color: #3498db;
  position: absolute;
  top: 50%;
  transform: translateY(-50%);

  /* Animation Properties */
  animation-name: slide;
  animation-duration: 4s;
  animation-timing-function: linear;
  animation-iteration-count: infinite;
}

/* Keyframes define the animation steps */
@keyframes slide {
  0% {
    left: 0;
  }

  100% {
    left: calc(100% - 50px);
  }
}
```

# **Question 7: Explain the difference between position: absolute; and position: fixed; in CSS.**

## **ANSWER :**

- The core difference between `position: absolute;` and `position: fixed;` in CSS lies in their reference point for positioning and their behavior when the user scrolls.

### **`position: absolute; `**

1. **`Reference Point :`**  It's positioned using its top, right, bottom, and left properties relative to its closest parent element that is itself positioned  (any ancestor with position set to relative, absolute, fixed, or sticky). If no such ancestor exists, it's relative to the `<body>` element.

2. **`Scrolling Behavior :`** Scrolls with the page content. Its position is static once set within its container.

3. **`Common Use :`**  Creating tooltips, dropdown menus, or complex layouts where a child element needs precise positioning within a specific parent container.




### **`position: fixed; `**

1. **`Reference Point :`**  It's positioned using top, right, bottom, and left properties relative to the browser window (viewport).

2. **`Scrolling Behavior :`** Because its position is tied to the viewport, it remains fixed even when the user scrolls the page.

3. **`Common Use :`** Creating sticky navigation bars, persistent headers/footers, or "Back to Top" buttons that must always be visible regardless of scrolling.

# **Question 8: Create a layered layout where a semi-transparent overlay covers an image using z-index. Provide the HTML and CSS code snippets.**

## **ANSWER :**
- A layered layout with a semi-transparent overlay covering an image using z-index can be achieved by placing both the image and the overlay inside a parent container and setting the overlay to have a higher z-index than the image.
- Here are the HTML and CSS code snippets to create this effect.

`HTML(index.html)`

```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Document</title>
    <link rel="stylesheet" href="style.css" />
  </head>
  <body>
    <div class="image-container">
      <img
        src="https://images.unsplash.com/photo-1759403117912-94e07f918cf3?w=600&auto=format&fit=crop&q=60&ixlib=rb-4.1.0&ixid=M3wxMjA3fDB8MHxmZWF0dXJlZC1waG90b3MtZmVlZHwzfHx8ZW58MHx8fHx8"
        alt="If img is not showing please insert a new link from internet "
      />

      <div class="overlay">
        <h2>Welcome to the View</h2>
        <p>This is a semi-transparent overlay sitting on top of the image.</p>
      </div>
    </div>
  </body>
</html>
```

`CSS(style.css)`

```
* {
  padding: 0;
  margin: 0;
  box-sizing: border-box;
}
body {
  font-family: Arial, sans-serif;
  display: flex;
  justify-content: center;
  align-items: center;
  min-height: 100vh;
  margin: 0;
  background-color: #f4f4f4;
}

.image-container {
  position: relative; /* This is crucial! It makes the overlay position relative to this container. */
  width: 600px;
  height: 400px;
  overflow: hidden;
  box-shadow: 0 4px 8px rgba(0, 0, 0, 0.2);
}

.image-container img {
  width: 100%;
  height: 100%;
  display: block; /* Removes extra space below the image */
  object-fit: cover; /* Ensures the image covers the container without distortion */
  position: relative; /* Give the image a z-index as well, though not strictly necessary here */
  z-index: 1; /* Image is on a lower layer */
}
img[alt] {
  color: red;
  font-size: 35px;
}
.overlay {
  position: absolute; /* Position the overlay relative to its positioned parent (.image-container) */
  top: 0;
  left: 0;
  width: 100%;
  height: 100%;
  background-color: rgba(0, 0, 0, 0.6);
  color: white;
  display: flex;
  flex-direction: column;
  justify-content: center;
  align-items: center;
  text-align: center;
  padding: 20px;
  box-sizing: border-box;
  z-index: 2;
}

.overlay h2 {
  margin-top: 0;
  font-size: 2em;
}

.overlay p {
  font-size: 1.1em;
  line-height: 1.5;
}
```
