# **`Question 1 `: In simple words, what are React Hooks? Create a small component that shows a button. When the button is clicked, toggle the text between “On” and “Off”.**

---

## **`Answer -->`**


### Understanding React Hooks and Implementation

- **Definition:**
**React Hooks** are special JavaScript functions introduced in React 16.8 that allow developers to "hook into" React features, such as state management and lifecycle methods, from within functional components. Before Hooks, these features were only available in Class components. Hooks enable a more modular and readable way to write complex components.


### Implementation: The Toggle Button

- To implement this, we use the `useState` hook, which allows the component to "remember" its internal state (the current status of the button).

**`Code Demo`**
```javascript
import React, { useState } from 'react';

/**
 * ToggleButton Component
 * Uses the useState hook to manage the 'isOn' boolean state.
 */
const ToggleButton = () => {
  // Declare a state variable 'isOn' and a function to update it 'setIsOn'
  const [isOn, setIsOn] = useState(false);

  // Toggle function
  const handleToggle = () => {
    setIsOn(!isOn); // Flips the boolean value
  };

  return (
    <div className="toggle-container">
      <button onClick={handleToggle}>
        {isOn ? "On" : "Off"}
      </button>
    </div>
  );
};

export default ToggleButton;

```


### Technical Breakdown

1. **`useState(false)`:** This hook initializes the state. The argument `false` is the initial value. It returns a pair: the current state value (`isOn`) and a function that lets you update it (`setIsOn`).
2. **Conditional Rendering:** Inside the button, we use a **ternary operator** `{isOn ? "On" : "Off"}`. This is a common React pattern where we dynamically determine the UI based on the current state.
3. **State Update:** The `handleToggle` function calls `setIsOn(!isOn)`. When this function is called, React schedules a re-render of the component with the new state value.


### Why Hooks are Advantageous

* **Simplification:** They remove the need for boilerplate code like `this` binding and class constructors.
* **Logic Reuse:** Custom hooks allow developers to extract stateful logic from a component and reuse it across multiple components without changing the hierarchy.
* **Performance:** Functional components with hooks are generally lighter for the React engine to process than class components.

# **`Question 2 `: Make a counter with a button. Whenever the count changes, log “Count changed” using useEffect.**

---

## **`Answer -->`**

###  Implementing Side Effects with `useEffect`

- Create a functional component that maintains a counter state and uses the `useEffect` hook to log a message to the console every time the count updates.

### Implementation (JavaScript/JSX)


**`Code Demo`**

```javascript
import React, { useState, useEffect } from 'react';

/**
 * Counter Component
 * Demonstrates state management and side effects.
 */
const Counter = () => {
  const [count, setCount] = useState(0);

  // The useEffect hook monitors the 'count' dependency
  useEffect(() => {
    console.log("Count changed");
  }, [count]); // Dependency array: Effect runs only when 'count' changes

  return (
    <div className="counter-container">
      <h2>Counter: {count}</h2>
      <button onClick={() => setCount(count + 1)}>
        Increment
      </button>
    </div>
  );
};

export default Counter;

```


### Technical Breakdown

#### 1. The `useEffect` Hook

- `useEffect` is used for "side effects"—actions that reach outside the component, such as data fetching, manual DOM manipulation, or logging.

* **The Dependency Array:** The second argument `[count]` is critical. It tells React to re-run the effect **only** when the value of `count` changes.
* **If the array were empty `[]`:** The effect would run only once, immediately after the initial mount.
* **If the array were omitted:** The effect would run after every single render of the component.

#### 2. Synchronization

- The `useEffect` hook ensures that the log message and the UI stay in sync. Because React schedules updates asynchronously, `useEffect` triggers **after** the browser has finished painting the screen, ensuring that the effect doesn't block the rendering process.

#### 3. State Management

- The `setCount` function from the `useState` hook triggers a re-render. When the component re-renders, React checks if the values in the dependency array have changed compared to the previous render. Since the `count` state has been updated, React triggers the `useEffect` callback.

# **`Question 3 `: Write a component that fetches users from https://jsonplaceholder.typicode.com/users only once on mount andshows how many users were loaded.**

---

## **`Answer -->`**

### Fetching Data with `useEffect`

- Create a functional component that fetches data from a public API (`https://jsonplaceholder.typicode.com/users`) exactly once when the component mounts and displays the total count of users fetched.


### Implementation (JavaScript/JSX)

**`Code Demo`**
```javascript
import React, { useState, useEffect } from 'react';

/**
 * UserCountFetcher Component
 * Fetches user data on mount and displays the count.
 */
const UserCountFetcher = () => {
  const [userCount, setUserCount] = useState(0);
  const [loading, setLoading] = useState(true);

  useEffect(() => {
    // Define the async fetch function
    const fetchUsers = async () => {
      try {
        const response = await fetch('https://jsonplaceholder.typicode.com/users');
        const data = await response.json();
        
        // Update state with the length of the retrieved array
        setUserCount(data.length);
      } catch (error) {
        console.error("Error fetching users:", error);
      } finally {
        setLoading(false);
      }
    };

    // Trigger the function
    fetchUsers();
  }, []); // Empty dependency array: ensures this runs only ONCE on mount

  return (
    <div className="fetch-container">
      {loading ? (
        <p>Loading...</p>
      ) : (
        <p>Total users loaded: {userCount}</p>
      )}
    </div>
  );
};

export default UserCountFetcher;

```


### Technical Breakdown

#### 1. The Empty Dependency Array `[]`

- Passing an empty array as the second argument to `useEffect` is the standard way to mimic the `componentDidMount` lifecycle method in functional components. It tells React that this effect does not depend on any props or state, so it should run exactly once after the initial render.

#### 2. Asynchronous Logic

- We use `async/await` syntax inside the `useEffect`. Note that we cannot make the `useEffect` callback itself `async` (e.g., `useEffect(async () => {...})`), because `useEffect` must return either `undefined` or a cleanup function. Thus, we define an internal `async` function and invoke it.

#### 3. State Management for UI States

* **`loading` State:** Crucial for providing visual feedback while the network request is in flight.
* **`userCount` State:** Holds the data derived from the API response. React updates the UI automatically once this state is set.

#### 4. Error Handling

- The `try...catch...finally` block ensures that the component handles network failures gracefully and that the `loading` state is cleared regardless of whether the request succeeded or failed.

# **`Question 4 `: Create a component that starts a timer with setInterval to increase a “seconds” counter every second. Stop the timer when the component unmounts.**

---

## **`Answer -->`**


### Implementing Timers with `useEffect` and Cleanup

-  Design a functional component that uses `setInterval` to increment a counter every second. Critically, the component must clear the interval when it unmounts to prevent memory leaks.


### Implementation (JavaScript/JSX)

**`Code Demo`**
```javascript
import React, { useState, useEffect } from 'react';

/**
 * TimerComponent
 * Demonstrates interval management and cleanup.
 */
const TimerComponent = () => {
  const [seconds, setSeconds] = useState(0);

  useEffect(() => {
    // Set up the interval and store the ID
    const intervalId = setInterval(() => {
      // Use functional state update to avoid dependency on 'seconds'
      setSeconds((prev) => prev + 1);
    }, 1000);

    // Cleanup function: Triggered when the component unmounts
    return () => {
      clearInterval(intervalId);
    };
  }, []); // Empty dependency array ensures this runs once on mount

  return (
    <div className="timer-container">
      <p>Elapsed time: {seconds} seconds</p>
    </div>
  );
};

export default TimerComponent;

```


### Technical Breakdown

#### 1. The `setInterval` Lifecycle

- In React, `useEffect` is the primary tool for managing timers. By placing the `setInterval` inside `useEffect` with an empty dependency array (`[]`), we ensure the timer initializes exactly once when the component is mounted.

#### 2. The Cleanup Function

- The `useEffect` hook allows returning a function, known as the **cleanup function**. React runs this function right before the component is removed from the DOM (unmount).

* **`clearInterval(intervalId)`:** This is mandatory. If you do not clear the interval, the JavaScript engine will continue to execute the function in the background even after the component is gone, causing a **memory leak** and potential application crashes.

#### 3. Functional State Update

- Notice the usage of `setSeconds((prev) => prev + 1)`.

* This is a "functional update" pattern.
* By not including `seconds` in the `useEffect` dependency array, we avoid having to re-create the interval on every tick. This pattern ensures we always have access to the most recent state value without causing unnecessary re-runs of the effect.


### Summary of Component Lifecycle

1. **Mount:** The `useEffect` runs, creating the `setInterval` timer.
2. **Tick:** Every 1000ms, the state updates, causing a re-render.
3. **Unmount:** React executes the cleanup function, clearing the interval and freeing the memory.

# **`Question 5 `: Build a small form with a text input. If the input is empty, show “Please type your name”. If it has text, show “Hello, <name>”.**




---

## **`Answer -->`**

### Conditional Rendering in Form Inputs

- Develop a functional component that captures user input via a text field. The UI must dynamically update based on the input's state: display a prompt "Please type your name" when empty, or "Hello, [name]" when valid text is provided.

### Implementation (JavaScript/JSX)

**`Code Demo`**
```javascript
import React, { useState } from 'react';

/**
 * NameInput Component
 * Handles controlled input and dynamic conditional rendering.
 */
const NameInput = () => {
  const [name, setName] = useState('');

  const handleChange = (e) => {
    setName(e.target.value);
  };

  return (
    <div className="form-container">
      <input
        type="text"
        value={name}
        onChange={handleChange}
        placeholder="Enter your name"
      />
      
      <div className="greeting-output">
        {name.trim() === '' ? (
          <p>Please type your name</p>
        ) : (
          <p>Hello, {name}</p>
        )}
      </div>
    </div>
  );
};

export default NameInput;

```


### Technical Breakdown

#### 1. Controlled Component

- This is a **Controlled Component** pattern. The state (`name`) is the "source of truth" for the input field.

* The `value` prop is set to the state variable.
* The `onChange` handler updates the state on every keystroke.
This synchronization ensures that React always knows the current value of the input.

#### 2. Conditional Rendering

- We use a **ternary operator** `{condition ? true : false}` to decide which JSX to render.

* `.trim()` is used to ensure that users cannot bypass the "Please type your name" message by simply typing spaces.
* If the trimmed string is empty, the first condition executes; otherwise, the "Hello" message appears.

#### 3. Why this approach?

* **Declarative UI:** We don't manually reach into the DOM to change text. We define what the UI should look like based on the `name` state, and React handles the updates.
* **Predictability:** Because the input is controlled, it is trivial to add further validation or transformations (e.g., converting text to uppercase) inside the `handleChange` function.

# **`Question 6 `: You have a theme value (“light” or “dark”) that is needed by a deep child button. Show how to pass it without sending theme through every parent prop.**

---

## **`Answer -->`**


### Solving "Prop Drilling" with the Context API

- In a standard React application, passing data through multiple layers of components ("Prop Drilling") makes code hard to maintain. Demonstrate how to share a `theme` value globally to a deep child component using the **Context API**.


### Implementation (JavaScript/JSX)

#### 1. Creating and Providing the Context

- The parent component creates a context and wraps its children in a `Provider` component.

**`Code Demo`**
```javascript
import React, { createContext, useContext } from 'react';

// 1. Create a Context object
const ThemeContext = createContext('light');

const App = () => {
  return (
    // 2. Wrap the component tree with the Provider
    <ThemeContext.Provider value="dark">
      <Toolbar />
    </ThemeContext.Provider>
  );
};

```

#### 2. Consuming the Context in a Deep Child

- The deep child component consumes the value directly using the `useContext` hook, bypassing all intermediate parents.

**`Code Demo`**
```javascript
const ThemedButton = () => {
  // 3. Consume the theme value directly
  const theme = useContext(ThemeContext);
  
  return (
    <button style={{ background: theme === 'dark' ? '#333' : '#fff', color: theme === 'dark' ? '#fff' : '#000' }}>
      I am a {theme} button
    </button>
  );
};

// Intermediate component that does NOT need the theme prop
const Toolbar = () => <ThemedButton />;

```

### Technical Breakdown

#### 1. The Context API Mechanism

- The Context API provides a way to share values between components without having to explicitly pass a prop through every level of the tree.

* **`createContext`**: Initializes the context object.
* **`Provider`**: A component that accepts a `value` prop and makes it available to all descendant components, regardless of how deep they are.
* **`useContext`**: A hook that allows a functional component to subscribe to the context and access the current `value`.

#### 2. Solving Prop Drilling

- Prop drilling occurs when you pass data through components that don't need it just to reach a deep child. Context breaks this dependency, keeping the intermediate components (like `Toolbar` in the example above) clean and decoupled from the application's configuration logic.

#### 3. Best Practices

* **Performance:** Context is best for low-frequency updates, such as theme settings, user authentication status, or locale/language preferences.
* **Over-usage:** Avoid using Context for highly dynamic data that changes every millisecond, as every consumer of the context will re-render whenever the value updates. For high-frequency state, consider a library like Redux or Zustand.

# **`Question 7 `: Make a counter with two `buttons: “+1”` and `“-1”`. Use useReducer instead of useState.**

---

## **`Answer -->`**



### Managing Complex State with `useReducer`

- Create a counter component with "Increment (+1)" and "Decrement (-1)" functionality. While `useState` is sufficient for simple counters, `useReducer` is the preferred approach for managing state logic that involves multiple actions or transitions.


### Implementation (JavaScript/JSX)

**`Code Demo`**
```javascript
import React, { useReducer } from 'react';

/**
 * 1. Define the reducer function
 * It takes the current state and an action, then returns the new state.
 */
const counterReducer = (state, action) => {
  switch (action.type) {
    case 'INCREMENT':
      return { count: state.count + 1 };
    case 'DECREMENT':
      return { count: state.count - 1 };
    default:
      throw new Error('Unknown action type');
  }
};

const Counter = () => {
  // 2. Initialize useReducer
  const [state, dispatch] = useReducer(counterReducer, { count: 0 });

  return (
    <div className="counter-container">
      <h2>Count: {state.count}</h2>
      
      {/* 3. Dispatch actions to trigger state updates */}
      <button onClick={() => dispatch({ type: 'INCREMENT' })}>+1</button>
      <button onClick={() => dispatch({ type: 'DECREMENT' })}>-1</button>
    </div>
  );
};

export default Counter;

```


### Technical Breakdown

#### 1. The `useReducer` Pattern

- `useReducer` is conceptually similar to Redux. It centralizes state-management logic outside the component body.

* **State:** The current data (here, `{ count: 0 }`).
* **Dispatch:** A function used to send "actions" (objects that describe *what* happened) to the reducer.
* **Reducer:** A pure function that calculates the next state based on the current state and the received action.

#### 2. Why `useReducer` over `useState`?

* **Scalability:** If your state needs to handle more complex logic (e.g., resetting, multiplying, or complex object updates), `useReducer` keeps that logic organized in one place rather than spreading it across multiple `useState` setters.
* **Predictability:** Because state updates are limited to defined "action types," it becomes much easier to debug state transitions as your application grows.
* **Logic Separation:** The reducer function can be easily moved to a separate file or utility module, allowing for better unit testing of your state logic independently of the UI.

# **`Question 8 `: Given an array `nums = [1,2,3,4]` and a filter value min, show how to compute a filtered list only when nums or min change. Also pass a stable click handler to a child list item.**

---

## **`Answer -->`**


### Memoization and Reference Stability in React

- Efficiently compute a filtered list from an array (`nums`) based on a `min` threshold, ensuring re-calculation occurs **only** when those specific dependencies change. Additionally, ensure the click handler passed to child components maintains a stable reference to prevent unnecessary re-renders.


### Implementation (JavaScript/JSX)

**`Code Demo`**
```javascript
import React, { useState, useMemo, useCallback } from 'react';

/**
 * FilteredList Component
 * Demonstrates useMemo for expensive calculations and
 * useCallback for function reference stability.
 */
const FilteredList = ({ nums, min }) => {
  // 1. useMemo: Memoizes the result of the filtering calculation
  // Recalculates ONLY when 'nums' or 'min' change.
  const filteredList = useMemo(() => {
    console.log("Computing filtered list...");
    return nums.filter((n) => n >= min);
  }, [nums, min]);

  // 2. useCallback: Memoizes the function reference
  // Ensures the child component doesn't re-render due to a new function prop
  const handleItemClick = useCallback((num) => {
    console.log(`Clicked item: ${num}`);
  }, []);

  return (
    <ul>
      {filteredList.map((num) => (
        <ListItem
          key={num}
          value={num}
          onClick={handleItemClick}
        />
      ))}
    </ul>
  );
};

// Child Component
const ListItem = React.memo(({ value, onClick }) => {
  console.log(`Rendering Item: ${value}`);
  return <li onClick={() => onClick(value)}>{value}</li>;
});

```

### Technical Breakdown

#### 1. `useMemo` for Expensive Calculations

- The `useMemo` hook caches the return value of the filter logic. Without it, the filtering would execute on **every render** of the parent component, even if the result were identical. By specifying `[nums, min]` as dependencies, we inform React to keep the existing result in memory until one of those values changes.

#### 2. `useCallback` for Reference Stability

- In JavaScript, functions are objects. If you define a function inside a component, it is re-created every time the component renders.

* **The Problem:** Passing a fresh function definition as a prop triggers a re-render in children wrapped in `React.memo`, even if the actual data hasn't changed.
* **The Solution:** `useCallback` returns a memoized version of the function that only changes if its dependencies change. This provides a "stable" reference to the child component.

#### 3. `React.memo` for Optimization

- The `ListItem` is wrapped in `React.memo`. This is a higher-order component that performs a shallow comparison of props. Because we used `useCallback` for `onClick`, the `onClick` prop remains stable across renders, allowing `React.memo` to skip re-rendering `ListItem` entirely if the `value` prop hasn't changed.


### Summary of Performance Optimization

| Hook | Purpose | Benefit |
| --- | --- | --- |
| **`useMemo`** | Memoize calculation results | Prevents redundant heavy logic execution. |
| **`useCallback`** | Memoize function references | Prevents child components from unnecessary re-renders. |
| **`React.memo`** | Memoize component output | Prevents re-rendering based on prop shallow comparison. |

# **`Question 9 `: Create a controlled input for email. As the user types, show `“Email: <value>”`. Why do we keep the input’s value in state?**

---

## **`Answer -->`**



### Controlled Inputs and State Management

- Create a controlled input component for an email field that displays the typed value in real-time. Explain the architectural importance of maintaining this input's `value` in the component's state.


### Implementation (JavaScript/JSX)

**`Code Demo`**

```javascript
import React, { useState } from 'react';

/**
 * EmailInput Component
 * Demonstrates a controlled input pattern.
 */
const EmailInput = () => {
  const [email, setEmail] = useState('');

  const handleChange = (e) => {
    setEmail(e.target.value);
  };

  return (
    <div className="input-container">
      <input
        type="email"
        value={email} // Controlled: React dictates the value
        onChange={handleChange} // Controlled: React handles the change
        placeholder="Enter your email"
      />
      <p>Email: {email}</p>
    </div>
  );
};

export default EmailInput;

```


### Technical Breakdown: Why keep the value in state?

- In React, a **Controlled Input** is one where the form element's value is controlled by the React state. We keep the input's value in state for several key technical reasons:

#### 1. Single Source of Truth

- By binding the input to the state, the React state becomes the "Single Source of Truth." The DOM's internal state (what the browser sees) and the React component's state are always synchronized. This avoids "out-of-sync" bugs where the browser and your application logic might disagree on what the user has entered.

#### 2. Declarative Validation and Transformation

- When the value is in state, we can easily perform logic *before* updating the UI.

* **Validation:** We can check the email format in real-time (e.g., checking for the presence of `@` or `.com`) and show error messages conditionally.
* **Transformation:** We can force the input to lowercase or strip invalid characters (like spaces) during the `onChange` event, and the input will immediately reflect these changes.

#### 3. Programmatic Control

- Having the value in state allows us to manipulate the input from anywhere in the component logic. For example, if we need to clear the field after a successful form submission (`setEmail('')`), we can do so effortlessly. In an uncontrolled component, this would require direct, manual DOM manipulation, which is discouraged in React.

#### 4. Easy Integration with Other Hooks

- Because the value is just a standard React state variable, it can easily interact with other hooks:

* **`useEffect`:** You can trigger an API call to check if an email is already registered as the user types (with debouncing).
* **`useMemo`:** You can derive other values from the input state to be used elsewhere in your application tree.

# **`Question 10 `: Make a button that focuses an input using a ref. Return adjacent elements without extra wrapper nodes. In one sentence, say what React’s update process does.**

---

## **`Answer -->`**


###  Refs and Fragments in React

> Implement a component where a button click triggers focus on an input field using `useRef`. The component must return multiple adjacent elements without adding unnecessary `<div>` wrappers to the DOM.


### Implementation (JavaScript/JSX)

**`Code Demo`**
```javascript
import React, { useRef } from 'react';

/**
 * FocusInput Component
 * Demonstrates the use of refs for DOM access and Fragments for clean markup.
 */
const FocusInput = () => {
  // 1. Create a ref to store the input DOM element
  const inputRef = useRef(null);

  const handleFocus = () => {
    // 2. Access the native DOM node and call the focus method
    inputRef.current.focus();
  };

  return (
    // 3. React Fragment (<>...</>) allows returning adjacent elements without a wrapper div
    <>
      <input
        ref={inputRef}
        type="text"
        placeholder="Click button to focus..."
      />
      <button onClick={handleFocus}>
        Focus the Input
      </button>
    </>
  );
};

export default FocusInput;

```


### Technical Breakdown

#### 1. `useRef` Hook

> In React, we generally avoid direct DOM manipulation. However, for certain tasks like managing focus, text selection, or integrating with third-party libraries, we use the `useRef` hook.

* **Mechanism:** `useRef` returns a mutable object with a `.current` property. When we pass this object to an element’s `ref` attribute, React sets `.current` to the corresponding DOM node when the component mounts.

#### 2. React Fragments (`<> </>` or `<React.Fragment>`)

- Components must return a single root element. Usually, developers wrap multiple elements in a `<div>`, which can lead to "div soup" and break CSS layouts (like Flexbox or Grid).

* **Benefit:** Fragments allow you to group a list of children without adding extra nodes to the DOM tree, keeping the HTML structure clean and semantically correct.


### React's Update Process (Definition)

> React’s update process, known as **reconciliation**, involves comparing the current Virtual DOM with a new version generated after a state change, identifying the minimal set of differences, and batching updates to the real DOM to maximize performance.


### Summary Table

| Feature | Tool Used | Purpose |
| --- | --- | --- |
| **Direct DOM Access** | `useRef` | Focus, scroll, or manual measurements. |
| **Adjacent Elements** | `Fragment` | Avoid extra DOM wrapper nodes. |
| **UI Syncing** | Reconciliation | Efficiently updating the real DOM. |
