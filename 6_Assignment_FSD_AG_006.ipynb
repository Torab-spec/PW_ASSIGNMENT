# **`Question 1 `: Explain what the forEach method does in JavaScript arrays and write code that uses forEach to display each student's name from an array of student objects to the console**


<h2> <b><code>
// Array of student objects <br>
const students = [
{ name: "Rahul", age: 20 },
{ name: "Priya", age: 19 },
{ name: "Amit", age: 21 },
{ name: "Sneha", age: 20 }
];
</code>
</b></h2>

---

## **`Answer -->`**

 ### JavaScript `forEach` Method

- The `forEach` method is a built-in array method in JavaScript used to execute a provided function once for each element in an array. It is primarily used for **iterative operations** where you want to perform an action (such as logging data, updating the DOM, or calculating a value) for every item in the collection.

#### Key Characteristics:

* **Execution:** It iterates through the array from the first element to the last.
* **Return Value:** It always returns `undefined`; it does not create a new array.
* **Syntax:** `array.forEach(callback(element, index, array))`


### Implementation Example

- To display each student's name from the provided array of objects, you can use the `forEach` method as follows:
<br>

**`JavaScript Code`**

```javascript
// Array of student objects
const students = [
  { name: "Rahul", age: 20 },
  { name: "Priya", age: 19 },
  { name: "Amit", age: 21 },
  { name: "Sneha", age: 20 }
];

// Using forEach to iterate and display names
students.forEach((student) => {
  console.log(student.name);
});

```

**Explanation of the code:**

1. We invoke the `forEach` method on the `students` array.
2. We pass an arrow function as a callback. This function receives the current object (`student`) from the array during each iteration.
3. Inside the block, we access the `name` property of the `student` object and print it to the console.

# **`Question 2 `: You have an array of product prices in rupees. Write code using the map method to create a new array that contains the same prices but with 18% GST added to each price.**

<h2>
<b>
<code>
const originalPrices = [100, 250, 500, 750, 1000];
</code>
</b>
</h2>

---

## **`Answer -->`**


### JavaScript `map` Method

- The `map` method is a powerful array transformation tool in JavaScript. Unlike `forEach`, which is used for side effects, `map` creates and returns a **new array** by applying a provided callback function to every element in the original array. This makes it ideal for data transformation while keeping the original data immutable.

#### Key Characteristics:

* **Immutability:** It does not modify the original array.
* **Transformation:** It constructs a new array of the same length as the original, containing the results of the transformation function.
* **Syntax:** `const newArray = array.map(callback(element, index, array))`


### Implementation Example

- To calculate the price with 18% GST ($18\%$ or $0.18$) added, we multiply the original price by $1.18$.

<br>

**`JavaScript Code`**

```javascript
// Array of product prices in rupees
const originalPrices = [100, 250, 500, 750, 1000];

// Using map to create a new array with 18% GST added
const pricesWithGST = originalPrices.map((price) => {
  return price * 1.18;
});

console.log(pricesWithGST);
// Output: [118, 295, 590, 885, 1180]

```

**Explanation of the code:**

1. We call the `.map()` method on the `originalPrices` array.
2. The arrow function takes each `price` and returns the calculation `price * 1.18`.
3. The `map` method collects these returned values into a new array, `pricesWithGST`, leaving the `originalPrices` array unchanged.

# **`Question 3 `: A school wants to find all students who scored more than 80 marks in theirexam. The following code has an error that prevents it from working correctly. Fix thecode and explain what was wrong.**


<h2>
<b>
<code>
const students = [ <br>
 { name: "Raj", marks: 85 },
 <br>
 { name: "Sita", marks: 92 },
  <br>
 { name: "Vikram", marks: 78 },
 <br>
 { name: "Meera", marks: 88 } <br>
]; <br>
const topStudents = students.filter(student => { <br>
 student.marks > 80; <br>
});
</code>
</b>
</h2>

---

## **`Answer -->`**

### Solution: JavaScript `filter` Method

- The `filter` method creates a new array containing only the elements that pass a specific "test" implemented by the provided callback function. For the test to work, the callback must explicitly **return** a truthy value (`true`) for elements that should be included in the new array.

#### The Error

- In the provided code, the arrow function uses a block body (enclosed in `{}`) but lacks a `return` statement. In JavaScript, if an arrow function has a block body, it does not implicitly return the result of the last expression. Consequently, the function returns `undefined` for every iteration, and the `filter` method interprets this as `false`, resulting in an empty array.


### Corrected Code

There are two common ways to fix this:

**Option 1: Using an explicit `return` statement**

<br>

**`JavaScript Code`**
```javascript
const topStudents = students.filter(student => {
  return student.marks > 80;
});

```

**Option 2: Using implicit return (most concise)**
If the function body is a single expression, you can omit the curly braces and the `return` keyword.


<br>

**`JavaScript Code`**
```javascript
const topStudents = students.filter(student => student.marks > 80);

```

### Implementation Example


<br>

**`JavaScript Code`**

```javascript
const students = [
  { name: "Raj", marks: 85 },
  { name: "Sita", marks: 92 },
  { name: "Vikram", marks: 78 },
  { name: "Meera", marks: 88 }
];

// Corrected filter logic
const topStudents = students.filter(student => student.marks > 80);

console.log(topStudents);
// Output: [ { name: "Raj", marks: 85 }, { name: "Sita", marks: 92 }, { name: "Meera", marks: 88 } ]

```

**Key takeaway:** When using `.filter()`, the callback function must return a boolean condition. If you use curly braces `{}` in an arrow function, you must include the `return` keyword, otherwise, the result will always be `undefined`.

# **`Question 4 `: Write code using the find method to locate a book with the title "Harry Potter" from an array of book objects. If the book is found display its author name, otherwise display "Book not found".**



<h2>
<b>
<code>
// Array of book objects
const books = [ <br>
{ title: "Gitanjali", author: "Rabindranath Tagore" },
<br>
{ title: "Harry Potter", author: "J.K. Rowling" },
<br>
{ title: "The God of Small Things", author: "Arundhati Roy" },
<br>
{ title: "Midnight's Children", author: "Salman Rushdie" }
<br>
];
</code>
</b>
</h2>

---

## **`Answer -->`**


### JavaScript `find` Method

- The `find` method is used to retrieve the **first element** in an array that satisfies a provided testing function. If no elements satisfy the testing function, `find` returns `undefined`. Unlike `filter`, which returns an array of all matches, `find` returns a single object or value, making it ideal for looking up specific items.

#### Key Characteristics:

* **Efficiency:** It stops iterating as soon as the first matching element is found.
* **Return Value:** Returns the element itself (or `undefined` if not found).
* **Syntax:** `const result = array.find(callback(element))`

### Implementation Example

To locate the book "Harry Potter", we use the `find` method and then verify the result using a conditional statement.

<br>

**`JavaScript Code`**

```javascript
// Array of book objects
const books = [
  { title: "Gitanjali", author: "Rabindranath Tagore" },
  { title: "Harry Potter", author: "J.K. Rowling" },
  { title: "The God of Small Things", author: "Arundhati Roy" },
  { title: "Midnight's Children", author: "Salman Rushdie" }
];

// Use find to locate the book object
const foundBook = books.find(book => book.title === "Harry Potter");

// Display the author name or an error message
if (foundBook) {
  console.log(foundBook.author);
} else {
  console.log("Book not found");
}

```

**Explanation of the code:**

1. **Search Logic:** `books.find(...)` iterates through the array and returns the first object where the `title` property strictly equals "Harry Potter".
2. **Conditional Check:**
* If `foundBook` holds an object, the condition evaluates to `true`, and we access `foundBook.author`.
* If the book is not present, `find` returns `undefined`, which is "falsy" in JavaScript, triggering the `else` block to print "Book not found".

<h2>
<b>
<code>
// Array of employee objects <br>
const employees = [<br>
 { name: "Rajesh", salary: 45000 },<br>
 { name: "Priya", salary: 55000 },<br>
 { name: "Amit", salary: 40000 },<br>
 { name: "Kavya", salary: 60000 }<br>

---

## **`Answer -->`**


### JavaScript `sort` Method

- The `sort` method in JavaScript is used to arrange the elements of an array. By default, it sorts elements as strings in lexicographical (alphabetical) order. Because of this default behavior, sorting numbers or objects requires a **compare function** to determine the correct order.

#### How the Compare Function Works

- To sort complex structures like objects, you provide a callback function that takes two elements (`a` and `b`) and returns:

* **A negative value:** If `a` should come before `b`.
* **A positive value:** If `b` should come before `a`.
* **Zero:** If the elements are considered equal.


### Implementation Example

- To sort the employee objects by salary in **descending order** (highest to lowest), you subtract the `salary` of the first element (`a`) from the second element (`b`).

<br>

**`JavaScript Code`**

```javascript
// Array of employee objects
const employees = [
  { name: "Rajesh", salary: 45000 },
  { name: "Priya", salary: 55000 },
  { name: "Amit", salary: 40000 },
  { name: "Kavya", salary: 60000 }
];

// Sort employees by salary in descending order
employees.sort((a, b) => b.salary - a.salary);

console.log(employees);

```

#### Explanation of the Code:

1. **Callback Function:** `(a, b) => b.salary - a.salary` is the comparison logic.
2. **Descending Logic:** By calculating `b.salary - a.salary`, if `b.salary` is higher, the result is positive, telling the `sort` method to place `b` before `a`.
3. **In-place Mutation:** It is important to note that the `sort` method modifies the original `employees` array directly rather than creating a new one.

# **``Question 6 ``: Write code using the reduce method to calculate the total value of all items in a shopping cart. Each item has a price and quantity. Also find the most expensive single item in the cart.**


<h2>
<b>
<code>
// Shopping cart with items <br>
const cart = [ <br>
{ name: "Laptop", price: 50000, quantity: 1 },<br>
{ name: "Mouse", price: 500, quantity: 2 },<br>
{ name: "Keyboard", price: 1500, quantity: 1 },<br>
{ name: "Monitor", price: 15000, quantity: 1 }<br>
];
</code>
</b>
</h2>

---

## **`Answer -->`**


### JavaScript `reduce` Method

- The `reduce` method is a powerful tool used to boil down an array into a single value, such as a sum, an object, or a maximum value. It executes a "reducer" callback function on each element of the array, resulting in a single output value.

#### How it Works:

* **Accumulator:** Stores the running result of the operations.
* **Current Value:** The current element being processed in the array.
* **Initial Value:** An optional value passed to the `reduce` function that sets the starting point for the accumulator.

### Implementation Example

- We can use `reduce` to calculate the total cost (price $\times$ quantity) and to compare prices to find the most expensive item.


<br>

**`JavaScript Code`**

```javascript
// Shopping cart with items
const cart = [
  { name: "Laptop", price: 50000, quantity: 1 },
  { name: "Mouse", price: 500, quantity: 2 },
  { name: "Keyboard", price: 1500, quantity: 1 },
  { name: "Monitor", price: 15000, quantity: 1 }
];

// 1. Calculate the total value of all items
const totalValue = cart.reduce((total, item) => {
  return total + (item.price * item.quantity);
}, 0);

// 2. Find the most expensive single item
const mostExpensive = cart.reduce((prev, current) => {
  return (prev.price > current.price) ? prev : current;
});

console.log("Total Value:", totalValue);
console.log("Most Expensive Item:", mostExpensive.name);

```

#### Explanation of the Code:

1. **Calculating Total Value:** We initialize the accumulator (`total`) at `0`. For every `item`, we add `item.price * item.quantity` to the `total`.
2. **Finding Most Expensive Item:** We compare the `price` of the current item with the previously "highest" item stored in the accumulator (`prev`). If `prev.price` is higher, we keep it; otherwise, we update the accumulator to the `current` item.

# **`Question 7 `: Explain what array destructuring is and write code that extracts the first two colors from an array of colors into separate variables using destructuring.**


<h2>
<b>
<code>
// Array of colors <br>
const colors = ["red", "blue", "green", "yellow", "purple"];
</code>
</b>
</h2>

---

## **`Answer -->`**

### Array Destructuring in JavaScript

- Array destructuring is a powerful syntax feature introduced in ES6 that allows you to unpack values from arrays into distinct, named variables using a syntax that mirrors array literals. Instead of manually accessing array elements by their numerical index (e.g., `colors[0]`), you can assign them to variables in a single, concise statement.

#### How it works:

* **Position-based:** Variables are assigned values based on their position in the array.
* **Declarative:** It makes code cleaner and more readable by avoiding repetitive index access.
* **Flexibility:** You can assign as many or as few elements as required, or use the "rest" operator to capture the remainder of the array.


### Implementation Example

- To extract the first two elements from the provided array, you place the desired variable names inside square brackets on the left side of the assignment operator.

<br>

**`JavaScript Code`**

```javascript
// Array of colors
const colors = ["red", "blue", "green", "yellow", "purple"];

// Extracting the first two colors using destructuring
const [firstColor, secondColor] = colors;

console.log(firstColor);  // Output: "red"
console.log(secondColor); // Output: "blue"

```

#### Explanation of the code:

1. **Declaration:** We define the source array `colors`.
2. **Destructuring Assignment:** By using `const [firstColor, secondColor] = colors;`, JavaScript automatically takes the element at index `0` and assigns it to `firstColor`, and the element at index `1` and assigns it to `secondColor`.
3. **Result:** The remaining elements in the array ("green", "yellow", "purple") are ignored in this assignment because no corresponding variables were provided for them.

# **``Question 8 ``: You have a student object with name, age, and marks properties. Write code using object destructuring to extract these values into separate variables and also rename the marks property to score while destructuring.**



<h2>
<b>
<code>
// Student object <br>
const student = { <br>
name: "Arjun",<br>
age: 20,<br>
marks: 85,<br>
city: "Mumbai"<br>
};

</code>
</b>
</h2>

---

## **`Answer -->`**


### Object Destructuring in JavaScript

- Object destructuring is an ES6 syntax that allows for the unpacking of properties from objects into distinct variables. Instead of accessing object properties via dot notation (e.g., `student.name`), you can define variables corresponding to the property names within curly braces `{}`.

#### Key Features:

* **Property Matching:** Variables are assigned based on the property keys of the object.
* **Aliasing:** You can rename a property during destructuring using the syntax `propertyName: aliasName`.
* **Flexibility:** You can extract only the specific properties required, ignoring any unused data.


### Implementation Example

To extract `name`, `age`, and `marks` (renamed to `score`) from the `student` object, use the following code:

<br>

**`JavaScript Code`**

```javascript
// Student object
const student = {
  name: "Arjun",
  age: 20,
  marks: 85,
  city: "Mumbai"
};

// Destructuring with renaming for the marks property
const { name, age, marks: score } = student;

console.log(name);  // Output: "Arjun"
console.log(age);   // Output: 20
console.log(score); // Output: 85

```

#### Explanation of the Code:

1. **Destructuring Block:** The syntax `{ name, age, marks: score }` instructs the engine to look for properties named `name`, `age`, and `marks` inside the `student` object.
2. **Aliasing:** By writing `marks: score`, you tell the engine to take the value of the `marks` property and assign it to a new variable called `score`.
3. **Implicit Extraction:** The variables `name` and `age` are created automatically using the object's keys, as no renaming was required for those specific fields.
4. **Ignored Properties:** The `city` property is omitted during this process, remaining untouched in the original object.

# **`Question 9 `: Write a function called calculateTotal that can accept any number of prices as arguments and returns their sum. Use the rest operator to handle the variable number of parameters.**


## **`Answer -->`**

---


### Using the Rest Parameter in JavaScript

- In JavaScript, the **rest parameter syntax** (`...`) allows a function to accept an indefinite number of arguments as an array. This is particularly useful when creating utility functions like `calculateTotal`, where the number of inputs is not known beforehand.

#### Key Mechanics:

* **Syntax:** `function functionName(...args)`
* **Behavior:** The rest parameter collects all remaining arguments passed into the function into a standard JavaScript array.
* **Flexibility:** Unlike the older `arguments` object, the rest parameter is a true array, allowing the use of built-in array methods like `reduce`, `map`, or `forEach` directly.


### Implementation Example

- The following implementation uses the `reduce` method to iterate over the collected array and compute the sum.

<br>

**`JavaScript Code`**

```javascript
/**
 * Calculates the total sum of any number of price arguments.
 * @param {...number} prices - A variable number of price values.
 * @returns {number} The sum of all prices.
 */
function calculateTotal(...prices) {
  return prices.reduce((accumulator, currentPrice) => {
    return accumulator + currentPrice;
  }, 0);
}

// Usage:
console.log(calculateTotal(100, 200, 50));    // Output: 350
console.log(calculateTotal(10, 20, 30, 40));  // Output: 100

```

#### Explanation of the Code:

1. **Rest Operator:** The `...prices` in the parameter list gathers all passed numeric arguments into a single array named `prices`.
2. **`reduce` Method:** We invoke `prices.reduce(...)` to perform the summation.
3. **Initial Value:** An initial value of `0` is provided as the second argument to `reduce`. This ensures that even if no prices are passed, the function returns `0` rather than throwing an error.

# **`Question 10 `: You have two arrays of student names and need to create a new array that contains all students from both arrays. Also, you have two objects representing a student's basic info and marks. Write code using the spread operator to solve both scenarios.**


<h2>
<b>
<code>
const classA = ["Rahul", "Priya", "Amit"];<br>
const classB = ["Sneha", "Vikram", "Pooja"];
</code>
</b>
</h2>

---

## **`Answer -->`**

### Using the Spread Operator (`...`) in JavaScript

- The **spread operator** (`...`) is an ES6 feature that allows an iterable (like an array) or an object to be expanded in places where zero or more arguments or elements are expected. It provides a clean, declarative way to merge arrays and combine or extend objects without using older, more verbose methods like `concat()` or `Object.assign()`.


### Implementation Example

#### 1. Merging Arrays

- To merge `classA` and `classB`, we expand the contents of both arrays into a new array literal.

<br>

**`JavaScript Code`**

```javascript
const classA = ["Rahul", "Priya", "Amit"];
const classB = ["Sneha", "Vikram", "Pooja"];

// Combining two arrays into one
const allStudents = [...classA, ...classB];

console.log(allStudents);
// Output: ["Rahul", "Priya", "Amit", "Sneha", "Vikram", "Pooja"]

```

#### 2. Merging Objects

The spread operator can also merge properties from multiple objects into a new object.

```javascript
const basicInfo = { name: "Arjun", age: 20 };
const marks = { subject: "Mathematics", score: 85 };

// Merging properties from both objects
const studentProfile = { ...basicInfo, ...marks };

console.log(studentProfile);
// Output: { name: "Arjun", age: 20, subject: "Mathematics", score: 85 }

```


### Technical Explanation

* **Array Merging:** By using `[...classA, ...classB]`, we create a shallow copy of all elements from `classA` followed by all elements from `classB` into a new, distinct array. This does not mutate the original arrays.
* **Object Merging:** When spreading objects `{...basicInfo, ...marks}`, JavaScript iterates over the enumerable properties of the source objects and copies them into the new object.
* **Conflict Resolution:** If both objects share a property key, the value from the **last** spread object will overwrite previous values. For example, `{...{a: 1}, ...{a: 2}}` results in `{a: 2}`.
